# Objective 4: numerical boundary condition (symmetry)

A numerical condition can be used to reduce the simulated domain in half provided there is symmetry that can be exploited. Suppose the situation is that there are two walls at a distance $x = \pm L$ from a center $x = 0$ with identical boundary conditions at either edge $x = \pm L$. Instead of simulating the full domain $x\in [-L,L]$ , and enforcing the two boundary conditions at each wall ($x = \pm L$), we can simulate the domain $x\in [0, L]$ where we replace the wall condition at $x = -L$ with a <i>symmetry boundary condition</i> at $x = 0$. <font color = "magenta">We henceforth refer to this simulated domain $\mathcal{D}_x = [0,L]$ as simply the <i>domain</i>, and we refer to the unsimulated domain $[-L, 0]$ as a <i>complement domain</i> $\mathcal{D}_x^C$</font>. A symmetry boundary condition communicates the effect of the complement density on the simulated domain so that simulating the domain with this condition gives a full picture of the "both sides" ($[-L,0]$ and $[0,L]$) as the interaction is factored in consistently. Such conditions are referred to as <i>numerical/</i> boundary conditions.

The form a symmetry boundary condition takes is now discussed. 

Such a condition enforces symmetry about $x = 0$, thus we can consider two half density functions $f$ and $f^C$ belonging to their respective domains, which together constitute the full density $f_{total}$ over the union of both domains:

$$f_{total}(t,x,v_x)\colon x\in [-L,L] = \mathcal{D}_x\cup\mathcal{D}_x^C, v_x\in\mathcal{D}_v$$

where the half densities are defined in:

$$f(t,x,v_x)\colon x\in\mathcal{D}_x = [0,L], v_x\in\mathcal{D}_v$$

$$f^C(t,x,v_x)\colon x\in\mathcal{D}_x^C = [-L,0], v_x\in\mathcal{D}_v$$

note that the velocity domain needs to be identical, else we would be introducing an asymmetry. The density definitions are such that $f = 0$ in $\mathcal{D}_x^C$ and $f^C = 0$ in $\mathcal{D}_x$. For clarity, thus the aforementioned total density 

$$f_{total}(t,x,v_x) = \begin{cases}
f(t,x, v_x) & x\in [0, L] \\
& \\
f^C(t, x, v_x) & x\in [-L,0] \\
& \\
0 & \text{otherwise}
\end{cases}\qquad \qquad \qquad\qquad (a)
$$

Actually, the domain being cut in half in $x$ translates to a symmetry condition $x \mapsto -x$, and $v \mapsto -v_x$, i.e.

$$f_{total}(t,x,v_x) = f_{total}(t,-x,-v_x), \qquad \text{symmetry about } x = 0, x\in [-L, L], v_x\in [-V, V]$$

For $x\in [0,L]$, we have $-x\in [-L,0]$, hence in terms of the half distributions $f$ and $f_C$ we have the corresponding equality:

$$\boxed{f(t,x,v_x) = f^C(t,-x,-v_x)}, \qquad \text{symmetry about } x = 0, \qquad \qquad (b)$$

We obtain this in more drawn out analysis by considering the definition $(a)$. If we consider $f_{total}$ at $x\in [0,L]$, then $f_{total}(t,x,v_x) = f(t,x,v_x)$. We then notice that $f_{total}$ at $-x\in [-L, 0]$ is $f_{total}(t,-x,v_x) = f^C(t,-x, v_x)$. We then notice that symmetry about $v_x$ requires $v_x \mapsto -v_x$, so we arrive again at condition $(b)$.

The above may momentarily seem to contradict pure symmetry only about $x = 0$, but is a consequence of this physical symmetry and how the form in phase space must take. Consider first a physical example that is consistent with symmetry about $x = 0$, then we will use this to motivate what this looks like in phase space. 

<u><b>Physical example</b></u>: We have symmetry about $x = 0$ with walls at $x = \pm L$ (can be replaced with periodic boundaries without consequence). Suppose we have an arbitrary density profile for a plasma system in the half plane $x \geq 0$ with purely right-moving velocity. Note that in this thought experiment, we are not thinking of the $x-v_x$ plane. In fact, we are not concerned with the detailed tracking of velocities of every particle, but instead would like to intuit a physical picture of a one dimensional line of plasma flowing towards the right wall. We only need to acknowledge that on this line the concentration of plasma particles generally varies and that each particle at the same $x$ can move with a different velocity than the others, though this cartoon does not rely on us knowing which particles move with what speed, only that they are all moving right.  Hence, the "distribution" we consider is actually the density $\int dv_x f = n(t,x)$. 

For simplicity, consider at time $t = 0$, $n(0,x)$ is localized between some range $\varepsilon \leq x \leq \varepsilon + \delta$, where $\varepsilon, \delta \in \mathbb{R}^+$, and assume we are far from $x = +L$ and not touching the boundary $x = 0$. At a later time, this density in general is redistributed according to the velocity differences at each $x$, but overall all points on the one-dimensional plasma move right towards the right wall. Since we have interest in simulating only this domain $x\in [0, L]$, the physical symmetry about $x = 0$ means there exists a density at $-\varepsilon - \delta \geq x \geq -\varepsilon$ at time $t = 0$ and moves left in the complement domain $[-L,0]$. This is agreeable to the idea of symmetry about $x = 0$. Of note, is that we required not just taking $x \mapsto -x$, but we also required the density to go from being right-moving to left-moving, thus it is understood the physical symmetry about $x = 0$ implies the phase space symmetry about both lines $x = 0$ and $v_x = 0$. Note that if we were to suggest that the plasma in the complement domain instead moves right, then we are not justified in using a symmetry condition as what happens on the left is not what is happening on the right with respect to the domain boundaries.

At time $t = 0$, we see that $f(t,x,v_x) = f(t,-x,-v_x)$. Consider the simple case for illustrative purposes, where the densities above are non-interacting, and are advected according to their velocity grid values. The density $f(t \geq 0, x, v_x)$ advects to the right where the distribution gets sheared according to the velocity increasing at higher ordinate values. Similarly, the density at $f^C(t\geq 0, -x, -v_x)$ advects left which takes the form of the reflection of the density $f(t,x,v_x)$ about the vertical mirror $x = 0$, and the horizontal mirror $v_x = 0$. Thus they both head towards their respective "walls" mirror along both axes, which accomplishes the intuitive physical symmetry about $x = 0$ where the "right density moves right, while the left density moves left." The reflection about both axes would hold even in the opposite situation (i.e. if $v_x < 0$ was considered).

Then, we see that condition $(b)$ holds. 

To cement a boundary condition that we can actually apply, we consider the following. For each particle that exits $x\in [0,L]$ with velocity $v_x < 0$, there exists a particle with equal but opposite velocity $v_x > 0$ entering $x\in [0,L]$ that originated from the "other side" (i.e. the complement domain $x\in [-L, 0]$). That is, we consider what are often called "ghost" particles (this author prefers the more indicative name "partner particles" or more endearingly, "friend particles"). All such exiting particles will have a partner particle that enter the simulated domain as they exit, thus this <i>friendly flux</i> (I'm joking, I won't be continuing to use terms like this) constitutes a source term of incoming particles for the simulated domain $x\in [0, L]$. We observe the source strength of this term is given by a continuity equation at $x = 0$:

$$(\Gamma^+_{\ell,(i,e)})(0) + (\Gamma^-_{r,(i,e)})(0) = 0, \qquad \qquad \text{symmetry boundary condition on flux}$$

where momentarily we use signatures $\ell$ and $r$ indicate domains on the left (complement domain, $[-L,0]$) and right (domain, $[L,0]$), respectively as it is more obvious. Equivalently, we have


$$\boxed{\Gamma^{-}_{\ell (i,e)}(0) = -\Gamma^{+}_{r,(i,e)}(0)}, \qquad \text{symmetry boundary condition on flux}$$

where

$$\Gamma^{\pm}_{i,e} = \Gamma^{\pm}_{i,e}(x) =  n_{(i,e)}(x)v_{x,(i,e)}$$

Note, we use delimiters $()$ to indicate an ordered tuple, i.e. we either choose $i$ on both quantities or $e$ on both. 

This is a true criterion, yet it is overly strict for the purposes of our implementation. We do not need to integrate to make this enforcement (i.e. to find the density $n_{e,i}$ in practice), we only need to replace each exiting density packet $f(t,x, v_x)$ with the density packet from the complement distribution $f^C(t,-x,-v_x)$. Thus, the weaker, yet equivalent, condition must be enforced:

The final equality is a statement that the particle $f(x^n, v^n_x)$ heading leftward ($v_x^n < 0$) over $t^n \rightarrow t^{n+1}$ from a prepoint $x^n\in [0,L]$ to a postpoint $x^{n+1}\in [-L,0]$ has a rightward moving ($-v_x^n > 0$) partner particle from the complement distribution $f^C$ starting at the image prepoint $-x^n\in [-L,0]$ which travels the same distance $|-x^{n+1} - (-x^n)| = |x^n - x^{n+1}|$ at that same originating time $t^n$ with postpoint at $-x^{n+1}\in [0, L]$.  Here, $x^{n+1} < 0$, $v_x < 0$, and since we do not directly have access to the (unsimulated) complement density $f^C$, we require translating $f^C$ in terms of $f$ in order to apply the statement above (and, also to later apply high order corrections to the flux $[Uf]^C$). 

We obtain this translation by considering the symmetry condition (eq. $(b)$), recall:

$$f(t,x,v_x) = f^C(t,-x,-v_x), \qquad \text{symmetry about } x = 0, \qquad \qquad (b, \text{ revisited})$$

Thus, we equate

$$f^C(t^n, -x^n, -v^n_x) = f(t^n, x^n, v^n_x)$$

so that, for all such qualifying prepoints, we realize we must increment:

$$f(t^{n+1}, -x^{n+1},-v^n_x) +\hspace{-.3em}= f(t^n, x^n, v^n_x)$$

In different words, the particle density packet $f(t^n, x^n, v_x^n)$ that exits has a partner particle packet $f^C(t^n,-x^n,-v_x^n)$ of the same size the enters the domain and settles at the postpoint $(-x^{n+1},-v_x)$. Thus, instead of dealing with the complement density, we can instead just reallocate the exiting density to this postpoint, as stated the increment update above.

This then is seen to provide the prescription which replaces all exiting particles with its partner particles that enter ($v_x \rightarrow -v_x$) and settle on the mirror postpoints, provided this point lies in the complement domain. That is, this is only for sufficiently low speeds (relative to their spatial prepoint) so that $x^{n+1}\in\mathcal{D}_x^C$. If the advection of a density at prepoint $x^n$ is pushed by a large enough speed $v_x$ such that it is pushed so far that it is pushed beyond $x = -L$ ($x^{n+1}\notin\mathcal{D}_x^C$, and of course $x^{n+1}\notin\mathcal{D}_x$), then the symmetry prescription above still is valid, but a boundary condition will need to be applied to its partner particle's postpoint location (e.g. a partner particle enters the domain and hits the <i>right</i> wall at $x = +L$). Thus, some wall boundary condition will need to be applied.

While a symmetry condition does not prohibit working out a symmetry condition over $2L$ (note the simulated domain would be of length $L$), this would require extra details to be include and it is unclear why a user would force this situation to happen rather than to just simulate the full domain as the motivation for periodic boundaries is to reduce the domain size in the first place. Reducing it further is not impossible, but seems forced the motivation is not clear. Hence, we take the case for symmetry boundaries being applied to nonperiodic cases and it the natural case, and we then require this to be encoded under <code>lib.boundaryconditions.nonperiodic</code>.

```python
def nonperiodic(f_old,
                Uf,
                z,
                vz,
                sim_params,
                charge,
                k = 0
                ):
    """orchestrates applying nonperiodic boundary conditions
    to the array w with total active grid points Nw. Nonperiodic
    boundary conditions require handling both left and right
    boundaries

    inputs:
    f_old -- (ndarray, ndim=2) density array
    z -- (instance) phase space variable being evolved

    outputs:
    f_old -- (ndarray, ndim=2) density with both left and right
             nonperiodic BCs enforced
    Uf -- (ndarray, ndim=2) high order fluxes with both left and right
             nonperiodic BCs enforced


    z returned (no changes) for symmetry with periodic routine above
    """
    # lower boundary
    f_old, Uf = eval(sim_params['BC'][z.str]['lower'] +
                           '_lower_boundary')(f_old, Uf,
                                              z.postpointmesh[k,:,:], z, vz,
                                              sim_params, charge)

    # upper boundary
    f_old, Uf = eval(sim_params['BC'][z.str]['upper'] +
                           '_upper_boundary')(f_old, Uf,
                                              z.postpointmesh[k,:,:], z, vz,
                                              sim_params, charge)
```

    sim_params['BC'][z.str]['lower'] = 'symmetric_lower_boundary' 
    
    or 
    
    sim_params['BC'][z.str]['upper'] = 'symmetric_upper_boundary'

We describe a complication here by sufficing to say that the efficient means of remapping is done through matrix addition and index slicing which does not have access to an increment operator <code>+=</code>, hence in our remapping, e.g. in <code>lib.convect.remap_assignment</code>:

    f_neg[ z.postpointmesh[0,:,:], vz.prepointmesh ] = f_old_ma + Uf_ma
    
if there are shared postpoints in <code>z.postpointmesh[0,:,:]</code> the right-hand side will not assign all such prepoints to each the same postpoints, only the last one "read-in" during the indexing process. Thus, we have previously made efforts to, for example, zero out any exiting particles and periodizing the postpoints to (1) ensure no two postpoints are shared to prevent densities from being overwritten, and (2) so that the postpoints they do map to have a contribution of zero added on. We acknowledge that overlapping postpoints is not a physical problem, but a conflict in how the implementation is set up. The second problem is that our velocity dependence is embedded in a flux $Uf$, hence it is not expected that there is not a relationshipo such as $Uf \rightarrow -Uf$ that exists, but rather a minus sign will infiltrate each weight on the derivative in each part that makes up $Uf$ (= sum of weighted derivatives).

### Setup and motivation of problem

We will consider the left boundary as being a symmetric boundary condition.

Suppose we have an $N_x \times N_{vx} = 5 \times 6$ grid, for the 2D grid $\{x_i, v_{xj}\}, i\in\mathbb{I}, j\in\mathbb{J}$. We assert the relationship $\Delta t = \Delta x$ for convenience here so that the CFL number which themselves are subclass objects of every phase space variable, e.g. $x$ here is the only relevant variable:

$$\mathcal{C}_{j} \equiv \text{x.CFL.numbers}[i,j] =  v_j\Delta t / \Delta x  = v_j\in\mathbb{Z}$$

In DECSKS, this is an attribute of the phase space variable: <code>x.CFL.numbers</code>.

#### Compute the postpoint locations $(i',j')$, i.e. $x_i \mapsto x_{i'}\colon i' = i + S, v_{j'} = \,?$

(note: when it is understood, we drop the diacritic and refer to the postpoint as simply $i$, which in general is our label for "rows" as well)

As usual, the CFL number is not in general an integer, $\mathcal{C}\in\mathbb{R}$, hence the prepoint density is allocated between to two contiguous $x$ postpoints which are appropriated in a way related to their CFL fractions $\alpha$ (a high order term is generated based on this). 

We store these in the 3D matrix <code>x.postpointmesh</code> where the first dimension stores nearest grid index to the true postpoint location (determined by $\mathcal{C}$) in its zeroeth entry, and the contiguous gridpoint to that in the same direction of travel is stored in the first entry.

### Remap implementation outline

Colloquially, we often use the name <code>k = x.postpointmesh</code> (also inside DECSKS), and <code>k1 = x.postpointmesh[0,:,:]</code> and <code>k2 = x.postpointmesh[1,:,:]</code> for the two postpoints which have the density be allocated per the fraction <code>x.CFL.frac</code> (note: for brevity we just refer to the CFL fraction, of course in practice we use a higher order term that ultimately depends on this fraction)

We currently engage the following stepthrough where we take an example of absorbing boundaries on both edges here for a moment just to illustrate how the routine works:

    lib.convect.remap_step: 

    # remap according to k1
    
        f_old -> lib.convect.remap_assignment(k1)

                 * lib.boundaryconditions.nonperiodic:
                    
                        - apply boundary condition on left (zero out density entries that exit)
                        - apply boundary condition on right (zero out density entries that exit)
                        - periodize k to account give every off-grid index an unshared on-grid index
                          (since all off-grid entries have their density zeroed out this amounts to 
                          remapping zero contribution to periodized, unique, postpoints. Thus, it is
                          of no consequence)
                          
                 * remap densities to f_new
                 
        f_k1 <- remap densities according to map k1


    # remap according to k2
    
        f_old -> lib.convect.remap_assignment(k2)

                 * lib.boundaryconditions.nonperiodic:
                    
                        - apply boundary condition on left (zero out density entries that exit)
                        - apply boundary condition on right (zero out density entries that exit)
                        - periodize k to account give every off-grid index an unshared on-grid index
                          (since all off-grid entries have their density zeroed out this amounts to 
                          remapping zero contribution to periodized, unique, postpoints. Thus, it is
                          of no consequence)
                          
                 * remap densities to f_new
                 
         f_k2 <- remap densities according to map k2
         
    return f_k1 + f_k2

## idea 2 description (used):

It has been realized that indexing with Boolean arrays can reduce the storage requirements and can be more efficient. In this way, we can reduce the above problem considerably by exploiting this. This stepthrough will amount to zeroing out complement entries in the density <code>f</code>, and <code>Uf</code>, and extracting those into a 1D array corresponding to the size of how many entries exited, i.e. <code>f_C.shape = (len(np.where(xpostpointmesh < 0))],)</code>. This is clearly of benefit given we reduce unnecessary handling of what are ultimately significantly sized arrays. 

We begin separating complement from non-complement objects:

$$\mathrm{f} = \left(\begin{array}{cccccc}
\color{blue}{\boxed{f_{0,0}}} & \color{red}{\boxed{f_{0,1}}} & f_{0,2} & f_{0,3} & \color{red}{f_{0,4}} & \color{blue}{f_{0,5}} \\
\boxed{f_{1,0}} & f_{1,1} & f_{1,2} & f_{1,3} & f_{1,4}& f_{1,5}\\

f_{3,0} & f_{3,1} & f_{3,2} & f_{3,3} & f_{3,4}& f_{3,5}\\
f_{4,0} & f_{4,1} & f_{4,2} & f_{4,3} & f_{4,4}& f_{4,5}
\end{array}\right)$$

The boxed entries exit and need to be replaced with information from the complement density, and we color the entries we know produce share postpoints upon symmetrizing (based on the above naive attempt at applying the symmetry boundary condition).

We partition the complement density and non-complement density as:

$$\mathrm{f\_C} = (\color{blue}{\boxed{f_{0,0}}},\color{red}{\boxed{f_{0,1}}}, \boxed{f_{1,0}}), \qquad \mathrm{f} = \left(\begin{array}{cccccc}
\boxed{0} & \boxed{0} & f_{0,2} & f_{0,3} & \color{red}{f_{0,4}} & \color{blue}{f_{0,5}} \\
\boxed{0} & f_{1,1} & f_{1,2} & f_{1,3} & f_{1,4}& f_{1,5}\\

f_{4,0} & f_{4,1} & f_{4,2} & f_{4,3} & f_{4,4}& f_{4,5}
\end{array}\right)$$

by index slicing with the boolean array <code>xpostpointmesh[0,:,:] < 0</code> on the density <code>f</code>, i.e.

    f_C = f[xpostpointmesh[0,:,:] < 0]
    
Similarly,

    f[xpostpointmesh[0,:,:] < 0] = 0
    
modifies the original density as above without falling back on an <code>np.where</code> function (comparable efficiency).

    [[1, 2, 3],

    is represented in the row-major order as:

    [1, 2, 3, 4, 5, 6]

    """

We continue now stepping through our example, by formalizing a bit what the symmetry condition will do based on the above. We start at the method <code>lib.convect.remap_step</code> which we paste without the docstring:

```python
def remap_step(
        sim_params,
        f_old,
        Uf,
        n,
        z,
        vz,
        charge
        ):

    f_k1 = remap_assignment(
        sim_params,
        f_old,
        Uf,
        z,
        vz,
        charge,
        index = 'nearest'    # remaps to nearest neighbor index
        )

    # remap to contiguous cell center
    f_k2 = remap_assignment(
        sim_params,
        f_old,
        Uf,
        z,
        vz,
        charge,
        index = 'contiguous'    # remaps to contiguous neighbor of above
        )

    f_remapped = f_k1 + f_k2

    return f_remapped
```

Each remapped density pertaining to k1 (nearest gridpoint to computed postpoint), and k2 (contiguous gridpoint to k1 in same direction of travel) is obtained by calling <code>remap_assignment</code> in succession. Below, we add comment lines to emphasize where the boundary condition is called

```python
def remap_assignment(
        sim_params,
        f_old,
        Uf,
        z,
        vz,
        charge,
        index = 'nearest'
        ):
    
    mask_neg =  (z.CFL.frac < 0)
    mask_pos = np.logical_not(mask_neg)

    if index == 'nearest':

        # copies are made as the boundary conditions modify f, Uf to achieve the required
        # result. After (when calling for contiguous mapping), we need to reuse the unchanged
        # f and Uf. Numpy arrays are passed by reference, so a copy is needed here
        f_BCs_applied = f_old.copy() # BCs will be applied to f_old
        Uf_BCs_applied = Uf.copy()

        # APPLY BOUNDARY CONDITIONS
        # e.g. for absorbing boundaries, we zero out all prepoint density entries
        # which exit the domain
        
        #************************************************************************
        #************************************************************************
        f_BCs_applied, Uf_BCs_applied, z = \
          eval(sim_params['boundarycondition_function_handle'][z.str])(
              f_BCs_applied, Uf_BCs_applied, z, vz, sim_params, charge, k = 0)
        #************************************************************************
        #************************************************************************
        
        f_old_ma = ma.array(f_BCs_applied)
        Uf_ma = ma.array(Uf_BCs_applied)
        f_pos, f_neg = ma.zeros(f_BCs_applied.shape), ma.zeros(f_BCs_applied.shape)

        # mask out negative values
        f_old_ma.mask = mask_neg
        Uf_ma.mask = mask_neg

        f_pos[ z.postpointmesh[0,:,:], vz.prepointmesh ] = f_old_ma - Uf_ma

        # mask out all positive values
        f_old_ma.mask = mask_pos
        Uf_ma.mask = mask_pos

        f_neg[ z.postpointmesh[0,:,:], vz.prepointmesh ] = f_old_ma + Uf_ma

    elif index == 'contiguous':

        # APPLY BOUNDARY CONDITIONS
        # e.g. for absorbing boundaries, we zero out all prepoint density entries
        # which exit the domain
        f_old, Uf, z = \
          eval(sim_params['boundarycondition_function_handle'][z.str])(
              f_old, Uf, z, vz, sim_params, charge, k = 1)

        # initialize masked array for flux Uf; mask out negative values
        f_pos, f_neg = ma.zeros(f_old.shape), ma.zeros(f_old.shape)
        Uf_ma = ma.array(Uf)
        Uf_ma.mask = mask_neg

        f_pos[ z.postpointmesh[1,:,:], vz.prepointmesh ] = Uf_ma

        # mask out all positive values
        Uf_ma.mask = mask_pos

        f_neg[ z.postpointmesh[1,:,:], vz.prepointmesh ] = -Uf_ma


    # "wherever there is negative data, assign f_neg, else assign f_pos
    f_new = np.where(mask_neg == True, f_neg.data, f_pos.data)

    return f_new
```

The starred area calls the boundary condition orchestrator. This correspondsn to an orchestrator <code>lib.boundaryconditions.periodic</code> or <code>lib.boundaryconditions.nonperiodic</code>. The symmetric is nonperiodic since one boundary is different than the other, hence the above calls (we have removed many comments that are in DECSKS for readability here)

```python
def nonperiodic(f_old,
                Uf,
                z,
                vz,
                sim_params,
                charge,
                k = 0
                ):
    # lower boundary
    f_old, Uf = eval(sim_params['BC'][z.str]['lower'] +
                           '_lower_boundary')(f_old, Uf,
                                              z.postpointmesh[k,:,:], z, vz,
                                              sim_params, charge)

    # upper boundary
    f_old, Uf = eval(sim_params['BC'][z.str]['upper'] +
                           '_upper_boundary')(f_old, Uf,
                                              z.postpointmesh[k,:,:], z, vz,
                                              sim_params, charge)


    z.postpointmesh[k,:,:] = periodize_postpointmesh(z.postpointmesh[k,:,:], z.N)
    # periodize mesh to give all off-grid points a unique on-grid location. These locations
    # will be mapped with zero density given the effect of the above boundary conditions

    return f_old, Uf, z
```

## Discerning the form of $[Uf]^C_{i,j}$

We know from symmetry that at each point $(x_i, v_{xj})$:

$$f^C(x_i, v_{xj}) = f(-x_i, -v_{xj})$$

In general,

$$f^C(x, v_x) = f(-x, -v_x)$$

Here we are considering a partner prepoint in the complement domain $x < 0 (x\notin\mathcal{D}_x),v_x\in\mathcal{D}_v > 0$. Differentiating this we have

$$\frac{\partial^{dn}f^C(x,v_x)}{\partial x^{dn}} = (-1)^{dn}\frac{\partial^{dn}f(-x,-v_x)}{\partial x^{dn}}$$

We have the prepoints $(x,v_x)$ above, for example, at $(x_{-1}, v_{x,5})$

$$\left.\frac{\partial^{dn}f^C(x,v_x)}{\partial x^{dn}}\right\rvert_{(x_{-1},v_{x,5})} = (-1)^{dn}\left.\frac{\partial^{dn}f(x,v_x)}{\partial x^{dn}}\right\rvert_{(x_{1},v_{x,0})}$$

or indicially,

$$\left.\frac{\partial^{dn}f^C(x_i,v_{x,j})}{\partial x^{dn}}\right\rvert_{(-1,5)} = (-1)^{dn}\left.\frac{\partial^{dn}f(x_i,v_{x,j})}{\partial x^{dn}}\right\rvert_{(1,0)}$$

in general, denote each entering partner prepoint in the complement domain as $(i^*, j^*)$ and its exiting partner with the corresponding indices $(i,j)$, then we have the general relation:

$$\left.\frac{\partial^{dn}f^C(x_i,v_{x,j})}{\partial x^{dn}}\right\rvert_{(i^*,j^*)} = (-1)^{dn}\left.\frac{\partial^{dn}f(x_i,v_{x,j})}{\partial x^{dn}}\right\rvert_{(i,j)}$$

where for a left symmetry boundary, we have

$$i^* = -i$$
$$j^* = N_{vx} - 1 - j$$

or in the most convenient form:

$$(d_{i^*,j^*}^{(dn)})^C = (-1)^{dn}d_{i,j}^{(dn)}$$

in the implementation, we assemble pairs $(i^*, j^*)$ in the array <code>partner_prepoints</code>, and the exiting prepoints $(i,j)$ in <code>exiting_prepoints</code>. Thus, we can access the needed derivatives easily.

A perspective that is useful here, rather than focusing on the partners entering, but on the exiting particles. That is, we "reposition" an exiting particle from $(i,j)$ to a postpoint positions $(k_1, j')$ and $(k_2, j')$ where $j' = N_{vx} - 1- j, k_1 = i + S, k_2 = i + S + 1$ ($v_x > 0$ for entering particles). We map the denseity $f(i,j)$, and compute a high order fraction $Uf \sim cd$, where we use <code>c[:, vx.N - 1 - j]</code> and <code>(-1) ** dn * d[dn,i,j]</code> which can be accomplished nicely as <code>(-np.ones(sim_params['N'])) ** np.arange(sim_params['N']) * d[:,i,j]</code>.

We now consider the question of equivalent between the complement fluxes and the fluxes already computed on-grid. It is obvious already, but we will show they are not "reusable" as is, and instead must be computed to some extent.

$$[Uf]^C_{i^*,j^*}  = \sum_{q} c^C_{q,j^*} (d_{i^*,j^*}^{(q)})^C$$

where

$$c^C_{q,j^*} = c_{q,j}, \quad j \text{ and } j^* \text{ are partners}$$

$$(d_{i^*,j^*}^{(q)})^C = (-1)^q d_{i,j}^{(q)}$$

Collecting the results:

$$[Uf]^C_{i^*,j^*}  = \sum_{q} (-1)^q c_{q,j} (d_{i,j}^{(q)}) = \underbrace{c_{0,j} (d_{i,j}^{(0)})}_{f_{i,j}} + (-1) c_{1,j} (d_{i,j}^{(1)}) +  c_{2,j} (d_{i,j}^{(2)}) + (-1)^3 c_{3,j} (d_{i,j}^{(3)})  +  \ldots $$

comparing with 

$$[Uf]_{i,j}  = \sum_{q}  c_{q,j} (d_{i,j}^{(q)}) = \underbrace{c_{0,j} (d_{i,j}^{(0)})}_{f_{i,j}} + c_{1,j} (d_{i,j}^{(1)}) +  c_{2,j} (d_{i,j}^{(2)}) + c_{3,j} (d_{i,j}^{(3)})  +  \ldots $$

Which shows that $[Uf]_{i^*,j^*}^C \neq [Uf]_{i,j} \neq -[Uf]_{i,j}$ or anything relatable in a simple way to the fluxes. <b>We cannot "translate" one flux into a complement flux</b>. Thus, we have to compute the complement fluxes using $c$ and $d$ according the above prescription. 

However, it is reassuring we do not need to recompute derivatives or correctors <code>c</code>, we can access the required values dictated by the partnership and compute the flux terms for only those partners that enter. In general, this number can be a fraction of the full density. Even in the extreme case where all particles from $f$ exit the domain as all particles from $f^C$ enter, we will not be computing parameters for the "full grid" evenstill. The fluxes from $f$ will be zeroed out, and only the complement fluxes will be computed. Thus, the savings in computational cost is still clear.

We also see there is no benefit to storing the fluxes from exiting prepoints as they have no use in the complement flux calculations. Thus, we have the following partition:


\begin{eqnarray*}
& \overbrace{\left(\begin{array}{cccccc}
