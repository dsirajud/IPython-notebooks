# DECSKS-25 -- Finite difference derivative schemes revisited on a cell-centered perspective

The finite difference schemes operating on node/vertex-centered grids. We now hav ea cell-centered grid, so that we must ensure our routines either work as is or we must develop the required translation to make them work.

We model our test problems off of notebook s-02:



We consider the infinitely differentiable function:

$$f(x) = \sin(\pi x) + 1 $$ 

which have derivatives ($n\in\mathbb{N}_1$) :

$$f^{(2n-1)}(x) = (-1)^{n+1} \pi^{2n - 1} \cos(\pi x)$$

and

$$f^{(2n)}(x) = (-1)^n \pi^{2n}\sin(\pi x)$$

Below, we plot the function as well as four normalized derivatives, the normalized derivatives of higher order are coincident with those plotted. Note, this is only to fit them on the same plot, for convergence we properly analyze the unnormalized forms.

We previous represented the grid as 

    x = np.linspace(ax, bx, Nx)
    
which includes the endpoints, and thus correspond to cell-edges (nodes/vertices in higher dimensional language. We now wish to represent the grid as cell-centered nodes:

    x, dx = np.linspace(ax, bx, Num = Nx, retstep = True, endpoint = False)
    x += 0.5*dx
        
This easily produces the cell-centered grid, which now differ from the edge nodes by an amount 1/2 * dx, and contains only interior points to the domain per the index slicing in the final step.

Encouragingly, we note that there was nothing edge-specific in our previous derivation in s-01 ("Generation of [FD schemes]"), thus we anticipate zero changes upon shallow thought.

## Verifying the linspace construction for cell-centered grids

```python
import numpy as np

ax, bx = 0., 1.
Nx = 5
x,dx = np.linspace(ax,bx,Nx, retstep = True, endpoint = False)
x += 0.5*dx
print x

# check with raw calc
aw, bw = 0., 1.
Nw = 5
dw = (bw - aw) / Nw

w = np.zeros(Nw)
for i in range(Nw):
    w[i] = aw + i*dw
    
w += 0.5*dx

print w == x
print dx == dw
```

```python
%reload_ext autoreload 
%autoreload 2 
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


x, dx = np.linspace(-1,1,num = 1000, retstep = True, endpoint = False)
x += 0.5*dx

f =  np.sin(np.pi*x) + 1
nmax = 4 # maximum derivative desired to be plotted

fig, ax = plt.figure(), plt.subplot(111)

ax.plot(x,f, linewidth = 2, label = '$f(x)$')
ax.hold('on')

for n in range(1,nmax + 1):
    if np.mod(n,2) == 1: # n is odd
        dnf = (-1) ** (n + 1) * np.pi ** (2*n - 1) * np.cos(np.pi * x)
        dnf /= np.pi ** (2*n - 1) # i.e. we plot normalized derivatives

    else: # n is even
        dnf = (-1) ** n * np.pi ** (2*n) * np.sin(np.pi * x)
        dnf /= np.pi ** (2*n) # i.e. we plot normalized derivatives

    ax.plot(x,dnf, linewidth = 2, linestyle = '--', label = r'$\pi^{-%i}\partial_x^{%i} f$' % (n,n) )

ax.hold('off')
frame = ax.get_position() # position of plot center

# shrink frame of the plot to make room for a legend on the right side
ax.set_position([frame.x0, frame.y0, 
                 frame.width * 0.8, frame.height])

# Place legend to the right of the shrunken frame
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5),
          fancybox=True, ncol=1)

plt.show()
```
