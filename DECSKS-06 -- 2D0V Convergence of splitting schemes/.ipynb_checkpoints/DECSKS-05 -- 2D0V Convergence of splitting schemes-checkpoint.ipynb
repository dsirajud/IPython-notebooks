# Convergence of splitting schemes

## A. Motivation

Five split schemes have been included in the implementation of DECSKS, where in the following, LTE is shorthand for local truncation error of the scheme:

<ol>
<li> LF2 ("leapfrog 2"), LTE order $O(\Delta t^3)$
<li> Y4  ("Yoshida 4"), LTE order $O(\Delta t^5)$
<li> O6-4, LTE order $O(\Delta t^5)$
<li> O11-6, LTE order $O(\Delta t^7)$
<li> O14-6, LTE order $O(\Delta t^7)$
</ol>

(See [1] for scheme particulars, e.g. splitting coefficients and ordering of composition)

Note, the schemes are labelled not after their LTE but up to the order they are correct to. This is different than the convention used in many discussions regarding solvers for PDEs (e.g. Discontinuous Galerkin, finite volume, convected scheme, $\ldots$). In a simulation, we quote a global error which is the error as measured at the final time step $ = LTE - 1$, hence an LF2 scheme should report an algebraic order of convergence of 2, since its LTE is of order 3 at each step and the global error (GE) is the accumulative error over $N_t = T / \Delta t$, where $T$ is a chosen simulation duration i.e. for a method above with LTE order $N+1$, the global error $\text{GE} = \text{LTE}\cdot \frac{T}{\Delta t} = O(\Delta t^{N+1})\cdot\frac{T}{\Delta t} = O(\Delta t^N)$

## B. Goals

To verify the convergence of all of the enumerated splitting methods. These schemes are implemented in DECSKS as keyword options in the DECSKS.lib.convect.scheme function call (which is read in from the input file DECSKS/etc/params.dat), thus switching between methods is simple. Verifying the order of accuracy is a strong check to ensure the methods have been implemented correctly.

## C. Setup

We consider a problem that permits splitting methods to be used (i.e. more than one variable in phase space), but whose analyticaly solution is known. We consider the same problem as [2], rigid body rotation with angular velocity $\omega$. This is the same problem used to demonstrate the splitting method in [1], though the order of the splitting methods was not proven numerically at that juncture. Thus, we set up the problem

$$v_x = -\omega y, \quad v_y = +\omega x$$

where $\omega = 2\pi$ so that in tmie $T = 1.0$ one revolution is achieved. The 2D advection equation gives the evolution of the density $f = f(t,x,y)$:

$$\partial_t f + 2\pi y \partial_x f - 2\pi x \partial_y f = 0$$

For the domain $(x,y)\in [-1,1]\times [-1,1]$ over a time $t\in [0,T]$. Since the domain evolving phase space variables $(x,y)$ yet the velocities are fixed in time, it is termed 2D0V. We use the flower distribution as in [1]

Running test cases on refined time grids, we record all results and assess algebraic convergence and report a summary plot. Rather than plotting versus the time, the abscissa reports a measure of computational expense: total number of time substeps throughout the entire simulation time $S\cdot N_{\Delta t}$ (one rotation, here, $S$ = number of time steps per time step, $N_{\Delta t} \equiv N_t$ is the number of time steps total). For example for each time step LF2 requires $S = 4$ substeps in total (at the cost of being more specific, code-wise, it is acknowledged that this can be reduced in cost. For the compromise of simple switching between methods as a single keyword argument, that efficiency has not been pursued insofar). In comparson, O6-4 requires $S = 13$ substeps per one time step. The number of substeps required per time step $S$ for each scheme is recorded in the table at the very bottom, along with the serial processor time required to complete the simulation.

Thus, we see all schemes converge to the required global error = $LTE - 1$.

The abscissa is the product of the number of substeps per time steps with the total number of time steps throughout the simulation, i.e. the total expense. In this way, schemes can be compared on any vertical line at any given abscissa value. Thus, this plot permits direct weighing of one scheme among the rest with regard to performance accuracy and expense. This allows more informed decisions about compromises in computational expense vs. accuracy. For a given simulation, this plot can be used as an aid for guidelines on which time splitting scheme might be employed. For example, while the $LF2$ scheme the least expensive of all, requiring at most 4 stages per one time step, it is outcompeted by $Y4$ if the number of total time substeps throughout the simulation $SN_t > 120$ or so, thus in such circmstances it would be wiser to use $Y4$ as it yields greater accuracy for the same amount of work. Similarly, $O6-4$ sigificantly outcompetes $Y4$ and $LF2$ for the same amount of work, as does O14-6. We point out to the reader that these comparisons have a lower bound on the number of time steps for all methods (see $N_t$ column from above output), for example we cannot make any claims about how $O14-6$ might "outperform" LF2 at lower values of $SN_t$, as there are such lower values computable, for $O14-6$, the lowest number of time steps is $N_t = 1$ for any method. $O14-6$ requires $SN_t = 29\cdot 1 = 29$ substeps total for the simulation, this is the minimum amount of computational expense that can be compared with other methods as there cannot be fewer than one time step throughout the simulation. Thus, we while we can compute the same for $LF2$ ($S = 4$ at $N_t = 1$), we find $SN_t = 3$ time substeps total for the simulation on $LF2$, but it is the only method that has such a computational expense, so there exists no comparison.
