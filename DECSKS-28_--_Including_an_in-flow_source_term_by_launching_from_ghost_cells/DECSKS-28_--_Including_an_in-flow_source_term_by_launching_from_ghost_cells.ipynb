# Including a source term by advecting in density from loaded ghost cells

Here, we consider a velocity dependent source distribution $S_{\alpha} = S_{\alpha}(v_x)$ that physically is an equilibrium distribution representative of a kinetic Bohm criterion. That is, we will first explore setting up a Bohm situation and aim to simulate the expected sheath. After, we can take it a step further and try generalized source functions to see if a sheath develops and further if the generalized Bohm criterion is upheld or not.

For specificity, we take Maxwellian distributions as in Rizzopolou (recall our normalized units, see a previous notebook if desired for how these work out, probably notebook s10 or so, titled "two species Vlasov simulations" or similar)

$$S_e (v_x) = \frac{1}{\sqrt{2\pi}}\exp\left(-\frac{1}{2}(v_x - v_{De})^2\right)$$

$$S_i (v_x) = \frac{1}{\sqrt{2\pi \tau / \mu}} \exp\left(-\frac{(v_x - v_{Di})^2}{2 \tau / \mu}\right)$$

where $\tau = T_i / T_e$, $\mu = m_i / m_e$, and the drift velocities $v_{D\alpha}$ are multiples of the electron thermal velocity. As worked out in a previous notebook, for convenience we note the Bohm velocity in our normalized units is given by $v_{Bohm} = 1 / \sqrt{\mu}$. For example, if we wished to have the distribution centered at two times the Bohm speed, we would put $v_{D\alpha} = 2\cdot 1 / \sqrt{\mu}$ and so on

Consider a domain $D_x = [0,L]$ where there is some Dirichlet condition at $x = 0$, and we want to simulate the effect of sources existing over the region $x \geq L$ flowing (leftward) into the simulation domain (edge region).

We could physically load $N_G$ such ghost cells and compute where they would advect to on-grid by reusing the advection algorithm. What is intended here to save both expense and storage is instead to compute a summary table of densities equipped with postpoint map that tells us this information without having to load extra ghost cells, and instead we remap by reading a precomputed table essentially.

That is, we regard the current time substep (can have different substeps in split schemes), and regard for every velocity (which is affiliated with a particular value of the velocity-dependent source $S_{\alpha}$) its range of postpoints in configuration from a thick layer of an unspecified, yet "enough", ghost cells so that all in-flowing particles are factored in. If this source region is constant, then we can create a table of densities along with postpoint maps that we can apply every time. That is, we can save a state and reapply it each time without recalculating. Since the advection in $x$ are known a priori, we can do this before the simulation starts and trivially apply the saved state each time.

We do this by regarding the following requirements when considering our velocity-only dependent source distributions:

<ol>
<li>for each velocity $v_{x,j}$ of the source distribution at every intended (known a priori) time substep, calculate the number $(N_{G})_j$ of $x$ ghost cells $\mathcal{G}$ within a "region of effect" (defined so that the source density packet for the furthest ghost cell outboard of the domain advects to a postpoint that exactly reaches the edge of the simulation domain). For brevity in subsequent communication, we define the set of singletons that contain the collection of all numbers of ghost cells for each $v_{x,j}$ as the "ghost cell range" $\mathcal{R}_G = \bigcup_j\{N_G\}_j$.
<li>for every ghost cell $\mathcal{G}_g$, calculate the CFL numbers $C := \frac{v_{x,j}(a_s\Delta t)}{\Delta x} = \lfloor C\rfloor  + \alpha$ to determine the number of cells they advect through a particular time step. Here, $a_s$ labels the splitting coefficient (fraction of unity) for the stage $s$ for the advection operator $\mathcal{X}^{a_s\Delta t}$.
<li>create a map of postpoints $k_{\ell}$ $(\ell = 1,2)$, for each density packet $S_{g,j}^{\alpha , n}$ at phase space prepoint $(g_i,j)$ (or <b>sometimes we write the index $g_i$ as simply $g$ for brevity</b>). Here we use the index $g_i = 0, 1, 2, \ldots G$ to enumerate the ghost cells $\{\mathcal{G}_g\}$, where the ghost cell $\mathcal{G}_0 = [x_{N_x-1/2}, x_{N_x+1/2}) = [x_{g_i-1/2}, x_{g_i+1/2})$ with cell-center at $x_{g_0} = x_{N_x}$. Thus, the correspondence between on-grid index $i$ and ghost index $g_i$ is 
$${}$$
$$i = N_x + g_i, \qquad \text{relationship between grid-based index } i \text{ and ghost-grid based index } g_i$$
$${}$$
above we mean "based" as in where $i = 0$ or $g_i = 0$ begins.

In general, $\mathcal{G}_g = [x_{g_i-1/2}, x_{g_i+1/2}) = [x_{N_x-1/2 + g_i}, x_{N_x+1/2 + g_i})$, and $\bigcup_{g_i = 0}^{g_i = N_G-1}\mathcal{G}_g = \mathcal{D}_G$, the domain of ghost cells.
<li><font color = "purple">Final step (only possible for constant source terms in time and position)</font>: summatively develop a 2D "layer" of source density $(\underline{\underline{f}}_{S\alpha})_{N_x\times N_{vx}}$that can be added on top of the distribution at each time step ($\underline{\underline{f}}^{n+1} +\hspace{-.3em}=\underline{\underline{f}}_{S\alpha}$)  that is the result of remapping all such $(g,j)$ to on-grid locations $(i,j)$. Additionally, for any such source particles that reach the opposite boundary ($b$), store the total number that do so in a scalar object
$${}$$
$$N_{gb} = \sum_{\substack{g \\ g\mapsto k_{\ell} \\ k_{\ell} \leq -1/2}}  \Delta x \sum_{\substack{j \\ v_{x,j} < 0}} \Delta v_x S^{\alpha , n}(x_g,v_{x,j}) \qquad \text{e.g. for a right in-flow boundary}$$
$${}$$

(if context can be found, future implementations can establish an object that records more information, e.g. prepoint $v_{x,g,j}$. Such a context is not of interest for current applications.)
</ol>

Some notes regarding the particulars follow:

We determine the ghost cell range, $\mathcal{R}_G = \mathcal{R}_G(v_{x,j}) = \bigcup_{j}(N_G)_j$ by identifying the number of ghost cells $(N_G)_j$ needed for each affiliated grid velocity $v_{x,j}$ depends only on the distance each ghost cell packet advects along its in-flow characteristic. By tracing each characteristic backward from a postpoint corresponding to the rightmost on-grid index $i_{max} = N_x-1$ we arrive at a ghost prepoint $g_{max} = (N_G)_j-1$ that is the furthest ghost cell outboard of the simulation domain that contributes the on-grid solution (i.e. defined such that a density packet stored at a ghost prepoint $g_i = g_{max} + 1$ advects to postpoints $k_1 = g_1$ and $k_2 = g_0$).

It is acknowledged that we are limited to determining the advection based on the nominal CFL value $\alpha \equiv \text{CFL}_j - \lfloor\text{CFL}_j\rfloor$, <b>so this in-flow source boundary condition is second-order</b> as we  have no (CS-style) means of correcting this to higher order. That is, for any advection we wish to correct, we add on order corrections to the flux of source particles $S^{\alpha} \equiv S_{\alpha}$ (we use whichever script is seen to unclutter any extra notation in a given context), $US$:

$$[US]_{g_i,j}^{\alpha n} = \alpha_{i,j}^n S^{\alpha , n}_{g_i,j} + c_{q=1,\forall i,j}d^{(q=1)}_{g_i,j} + c_{q=2,\forall g_i,j}d^{(q=2)}_{g_i,j}  + c_{q=3,\forall i,j}d^{(q=3)}_{g_i,j}  + \ldots + c_{q=N-1,\forall i,j}d^{(q=N-1)}_{g_i,j}  + O(\Delta x^{N+1})$$

or, 

$$[US]_{g_i,j}^{\alpha , n} = \alpha_{g_i,j}^n S^{\alpha ,n}_{g_i,j} + \sum_{q = 1}^{N-1} c_{q,\forall g_i,j}d^{(q)}_{g_i,j} + O(\Delta x^{N+1})$$

where as in previous mentions we use the notation $\forall i$ to represent that $c = c(v_x\Delta t)$, so the coefficients are the same for each $i$ which marks the configuration variable.

If we analyze just one velocity at a given position, we can write the above in brief (note: though this is not a good way to implement this in practice if electing to use pure NumPy/Python):

$$[US]_{g_i,j}^{\alpha , n} = \alpha_{i,j}^n S^{\alpha ,n}_{g_i,j} + \underline{c}_{q,\forall g_i,j}\cdot\underline{d}^{(q)}_{g_i,j} + O(\Delta x^{N+1})$$

Now for a position independent source term $S_{\alpha} = S_{\alpha}(v_{x})$, we have every $d^{(q)}_{g_i,j} \equiv 0$ for all $q \geq 1$ since $d^{(q)}_{g_i,j} \sim \partial_x^{(q)}S_{\alpha}(x,v_x) = \partial_x^{(q)}S_{\alpha}(v_x) = 0$.

Thus, at best we have:

$$[US]_{g_i,j}^n = \alpha_{i,j}^n S^{\alpha , n}_{g_i,j}$$

This only suggests infinite accuracy in the case of truly constant distribution functions. Indeed, there is zero error for a constant distribution that is advected and remapped to a discrete grid, for any diffusive losses are restored by contiguous density cell contributions in a semi-Lagrangian scheme (consider a simple 1D example with constant upwind velocity $v_x$ for a density $f = f_0 = \text{const}$ with periodic boundary conditions, there will be zero diffusion even on a computer since this setup is an infinite domain where $f_0$ is the same everywhere. By however much a particular cell's postpoint density is lacking as compared to $f_0$ is exactly made up for by the contiguous cell's postpoint density remap). In our case, if our ghost cell number is quite large as compared to the on-grid cell number, then we have a better estimate than if we have a smaller ghost cell number (domain is not close to infinitely large by any comparative measure).

## Determining $(N_G)_j$

Since the advection along characteristics is in general a non-integral number of cell widths, we have just one more layer of consideration as compared to the simple prescription presented above. For a given $j$ ($v_{x,j}$), we calculate:

$$\text{CFL}_j^n = \frac{v_{x,j}\Delta t}{\Delta x} = \lfloor \text{CFL}\rfloor + \alpha \qquad \text{general, floor rounds towards zero, not } -\infty$$

For a leftward moving source distribution, this is explicitly the ceiling function, so we may write the above as follows:

$$\text{CFL}_j^n = \frac{v_{x,j}\Delta t}{\Delta x} = \lceil \text{CFL}\rceil + \alpha \qquad \text{for a left-moving source}$$

For a right in-flow condition to have a minimum proportion allocated to the on-grid simulation domain, we must have the postpoint $k_2 = N_x-1$, note that $v_{x,j} < 0$, hence the $k_2$ postpoint marks the leftmost gridpoint a prepoint density is allocated to. If $k_2 = N_x-1$ then $k_{true} \in [k_2, k_1] = [N_x-1, N_x]$, where $k_1 = i + \lceil\text{CFL}\rceil = N_x + g_i + \lceil\text{CFL}\rceil$ we must also have that $k_1 = N_x \equiv g_0$. In general, the $k_{1}$ postpoint for the ghost cell is given by

$$N_x + g_i + \lceil \text{CFL}_j\rceil = k_{1}$$

(and the $k_2$ postpoint is $k_2 = k_1 - 1$)

We find the number of ghost cells $N_G$ needed by enforcing the scenario that the ghost prepoint $g$ corresponds to the furthest cell, $g = N_G-1$, and that it advects to the edge of the simulation domain so that while its postpoint $k_1 = N_x\notin \mathcal{D}_x$, we do have that $k_2 = N_x-1\in \mathcal{D}_x$. Thus, regarding $k_1 = N_x$ or $k_2 = N_x-1$ is the same circumstance as $g + \text{CFL}_j \equiv k_{true}\in [N_x-1, N_x]$ Thus, we rewrite the above statement as

$$N_x + (N_G - 1) + \lceil \text{CFL}_j\rceil = N_x \Rightarrow N_G = 1 - \lceil \text{CFL}_j\rceil, \qquad \text{Maximum number of ghost cells needed for a velocity } v_{x,j}$$

or 

$$\boxed{(N_G)_j = |\text{CFL}_j| + 1} \qquad \text{total number of ghost cells that contribute to the simulation domain}$$

Now that the total number of ghost cells is calculable, we figure prescriptions that indicate the remap summaries for all proportions $\chi_1$ and $\chi_2$ for every source term $S_{g_i,j}$. We learn how to do this by considering two examples, then generalizing.

<li>if $-[\alpha S]_{g_{i,crit}^L,j} > 1/2\cdot S_{g_{i,crit}^L,j}$, then $\chi_1$ and $\chi_2$ reach the left boundary

<li>$\chi_2 = -2[\alpha S]_{g_{i,crit}^L,j}$ does reach the left boundary
</ul>
"""

Since $g_0 > g_{i,crit}^L$, this checks out, no edge cases present.

next, we have

"all ghost cells $g^{L}_{i,crit}+1 \leq g_i < N_G-1$ have both proportions $\chi_1$ and $\chi_2$ mapped on-grid"

This translates to

$$g^{L}_{i,crit}+1 \leq g_i < N_G-1 \Rightarrow -1 + 1 \leq g_i < 4 - 1$$

or

$$0 \leq g_i < 3 \quad \text{have both proportions are mapped to the grid} \checkmark$$

This agrees with the above table in example (1a) where $S_{g_0}, S_{g_1}, S_{g_2}$ have both proportions mapped on-grid.

Next, we have

"if ghost cell $g_i = N_G-1$, then the $\chi_2$ proportion maps to $k_2 = N_x-1$ by definition of $N_G$."

which means if $g_i = 4 - 1 = 3$, i.e. $g_i = g_3$ then only the proportion $\chi_2$ is mapped. $\checkmark$, this also checks out.

Now, we can store the on-grid densities in a 2D array as mentioned (summatively including both proportions), and add them at each time substep. And, we can store the boundary mapped source densities in a scalar object and increase the wall charge (for example) by this amount each time (or simply not include them in the case of an left absorber).

The convenient part about this is that we relaize we just need to do this one time, which happens before the simulation, so we do not even need to be efficient about doing this (though we will, as it is easy to do so by reusing Cython methods already programmed). This reduces the cost of the in-flow source boundary condition significantly. The <b>last question is what to do about on-grid particles that flow right out of the domain? We "absorb" them</b>. This the same thing as just stopping to keep track of exiting characteristics after they move passed a boundary.

# In-flow right source boundary implementation:

    Before simulation starts
    
    (1) Initialize velocity dependent source in a 1D array, shape = (Nvx,)
    
    (2) for every velocity, follow the above prescription to determine remap arrays f_Se and f_Si: shape = (Nx, Nvx)
        store these in the sim_params
        
    (3) in this process, we construct the total number of particles that reach the left boundary N_Se and N_Si
        store these in sim_params
     

<u>Example: left floating electrode, right in-flow source alpha = e,i</u>

    During simulation (s is the stage number, s = 1, 2, ...)
            
    (1) left boundary is applied (e.g. floating electrode)
    
            if right boundary == in-flow source:
            
                sigma_left += NSalpha[s] * charge
                
             proceed as usual for a floating electrode boundary condition:
             
                 collect proportions and update charge: sigma_left += Nalpha * charge
                 
                 remove collected proportions from distribution function 

    (2) right boundary is applied 
    
        falpha+= f_Salpha[s,:,:]
                
        absorb particles that exit right-hand side.

Note we have chosen the ordering a first applying the left condition then the right because that is just the current setup. As can be seen above, since we have separate objects N_Salpha and f_Salpha, it is easy to interchange the steps at no consequence.

Or we can just as easily do the following:

    During simulation
    -----------------
    
    split scheme (advect in x, then vx, then x, ...)
    
        advect in x
            
            (1) left boundary is applied (e.g. floating electrode)
                           
                     proceed as usual for a floating electrode boundary condition:
                     
                         collect proportions and update charge: sigma_left += Nalpha * charge
                         
                         remove collected proportions from distribution function 
        
            (2) right boundary is applied
                        
                absorb particles that exit right-hand side.
                
        
            fe += f_Se[s,:,:] # s is the stage number of the split scheme, s = 1, 2, ...
            fi += f_Si[s,:,:]
            
            sigma_left +=  N_Si[s]
            sigma_left += -N_se[s]
        
        advect in vx
        
            ...
            
Where we have removed all the boundary behavior from any vlasov function routine passes, and tack them on after the evolution happens. Since we reduced the behavior of the source terms to addition, we can do this whenever we see is consistent and convenient. It is more convenient to not have to object handle more than necessary. We pursue this idea.

## sources.py

```python
import numpy as np
import DECSKS

def setup(x, vx, t, sim_params):

    # get source distributions
    Se = get_distribution(vx, source = 'maxwellian', vD = -1 / np.sqrt(sim_params['mu'])) # left-moving
    Si = get_distribution(vx, source = 'maxwellian', mu = sim_params['mu'], tau = 1.0, vD = -1 / np.sqrt(sim_params['mu'])) #left-moving

    # print out source selections
    print "--------------------------------------------------------------------------------"
    print "\nAn upper source term has been specified in params.dat. The following sources have been selected:"
    print "\nelectrons: Maxwellian with drift velocity vD = %s" % 'Bohm velocity'
    print "ions     : Maxwellian with drift velocity vD = %s, mu = %s, tau = %g\n" % ('Bohm velocity', 'Hydrogen', 1.0)
    print "--------------------------------------------------------------------------------"

    # create 1D data for relevant velocity values
    Se_data = np.ones(len(vx.gridvalues[vx.gridvalues < 0]))
    Si_data = np.ones(len(vx.gridvalues[vx.gridvalues < 0]))

    #    Se_data = Se(vx.gridvalues[vx.gridvalues < 0])
    #    Si_data = Si(vx.gridvalues[vx.gridvalues < 0])

    # init CFL numbers for ghost cells (GCFL) compute CFL numbers for all
    GCFL = SourceAdvection(vx, sim_params)
    GCFL.compute_numbers_for_all_stages(sim_params, x, vx, t)

    # compute remap array f_S and object N_S that tells how many source particles reach the lower boundary

    # electron source
    f_Se, N_Se  = determine_source_effect_on_grid(
        x, vx, Se_data, GCFL, sim_params
        )

    # ion source
    f_Si, N_Si  = determine_source_effect_on_grid(
        x, vx, Si_data, GCFL, sim_params
        )

    store_global_vars(sim_params, f_Se, f_Si, N_Se, N_Si)


#==============================================================================#
# supporting routines
#==============================================================================#

def store_global_vars(sim_params, f_Se, f_Si, N_Se, N_Si):

    sim_params['sources'] = {}
    sim_params['sources']['N_Se'] = N_Se
    sim_params['sources']['N_Si'] = N_Si

    sim_params['sources']['f_Se'] = f_Se
    sim_params['sources']['f_Si'] = f_Si

    return None

def get_distribution(vx, source = 'maxwellian', **kwargs):
    """
    Returns the function object for a specified source
    """

    def source_function(vx):

        if source == 'constant':

            return np.where(vx < 0)

        if source == 'maxwellian':

            # check inputs, provide normalization factors as compared to electron species, e.g. mu, tau
            if 'mu' in kwargs.keys():
                mu = kwargs['mu']
                assert ( type(mu) != str)
            else:
                mu = 1.0 # dummy value

            if 'tau' in kwargs.keys():
                 tau = kwargs['tau']
                 assert ( type(tau) != str)

            else:
                tau = 1.0 # dummy value

            if 'vD' in kwargs.keys():
                 vD = kwargs['vD']
                 assert ( type(vD) != str)

            else:
                print "\nCourtesy notice: no value for the drift velocity vD has been specified, using vD = 0.0"

            return 1 / np.sqrt(2*np.pi * tau / mu) * np.exp(-1/2. * (vx - vD) ** 2 / (tau / mu))

    return source_function


def determine_source_effect_on_grid(x,
                                    vx,
                                    S,
                                    GCFL,
                                    sim_params
                                    ):
    """
    create objects f_S and N_S that are the objects that capture
    the effect of a bulk source flowing in from the right

    inputs:
    x  -- (instance)
    vx -- (instance)
    S  -- (ndarray, ndim=1) source function data for all vx < 0, len(S) < vx.N
          only negative values
    GCFL -- (instance) ghost CFL numbers and parts for source terms
    sim_params -- (dict) simulation parameters

    outputs:
    f_S -- (ndarray, ndim=3), shape = (stages+1, x.N, vx.N) density array that is
           the contribution of the source particles to every [i,j] for
           a velocity dependent distribution function

           Note: while the dimensions span vx.N columns, it is usual that
           half of these entries would correspond to positive velocities
           and thus are empty (source particles with postive velocity cannot
           flow inward [left] to the on-grid domain). The array
           is dimensioned of equal size to the arrays fe and fi in DECSKS
           simulations so that straightforward addition can accomplish
           the remapping on top of the evolved distributions fe, fi

    N_S -- (ndarray, ndim = 1), shape = (stages+1,) the number of source
            particles that reach the lower boundary

    see notebook s28 for details
    """

    num_stages = sim_params['splitting']['number_of_stages']['a']

    # create an array is indexed as (s,i,j) where s labels the stage (>= 1)
    f_S = np.zeros((num_stages + 1, x.N, vx.N))

    # create a 1D array indexed by s, the stage (s >= 1; s = 0 entry is empty)
    N_S = np.zeros(num_stages + 1, dtype = int)

    # send to Cython routine
    f_S, N_S = DECSKS.lib.source_advection.create_postpoint_objects(S,
                                GCFL.int,
                                GCFL.frac,
                                num_stages,
                                x.N,
                                vx.N,
                                x.width,
                                vx.width,
                                sim_params
                                )

    return f_S, N_S

class SourceAdvection:

    def __init__(self, vx, sim_params):
        # incoming source from right-side, i.e. lefgoing velocity

        dim1 = (sim_params['splitting']['number_of_stages']['a'] + 1,) # one-tuple
        dim2 = (len(vx.gridvalues[vx.gridvalues < 0]),) # leftgoing only, one-tuple
        dims = dim1 + dim2

        self.numbers = np.zeros(dims)
        self.int     = np.zeros(dims, dtype = int)
        self.frac    = np.zeros(dims)

        # dims
        self.nrows    = dims[0] # access 0th entry of the one-tuple
        self.ncols    = dims[1] # access 1st entry of the one-tuple

        return None

    def compute_numbers_for_all_stages(self, sim_params, x, vx, t):
        """
        Essentially a copy of the routine from lib.domain
        the difference is the return is not 2D, but 1D with size
        corresponding to the velocity grid domain
        """

        splitting = sim_params['splitting']

        # recall stages start numbering at 1, s = 1, 2, ..., not zero
        # when we are accessing the information as stored in
        # the dict sim_params['splitting']
        for s in range(1, splitting['number_of_stages']['a'] + 1):
            split_coeff = splitting['a'][s]
            self.compute_numbers_for_stage_s(x, vx, split_coeff*t.width, s)

        return None

    def compute_numbers_for_stage_s(self, x, vx, dt, s):

        self.numbers[s,:] = vx.gridvalues[:self.ncols] * dt / x.width
        print "dt = %g" % dt
        print "x.width = %g" % x.width
        self.int[s,:]     = np.ceil(self.numbers[s,:]) # round towards zero
        print "cfl int = "
        print self.numbers[s,:]
        self.frac[s,:]    = self.numbers[s,:] - self.int[s,:]

        return None
```

## source_advection.pyx (Cython)

```python
import numpy as np
cimport numpy as np
cimport cython

DTYPE = np.float64
DTYPEINT = np.int64

ctypedef np.float64_t DTYPE_t
ctypedef np.int64_t DTYPEINT_t

@cython.boundscheck(False)
def create_postpoint_objects(np.ndarray[DTYPE_t, ndim=1] S,
                             np.ndarray[DTYPEINT_t, ndim=2] GCFL_int,
                             np.ndarray[DTYPE_t, ndim=2] GCFL_frac,
                             int num_stages,
                             int Nx,
                             int Nvx,
                             DTYPE_t xwidth,
                             DTYPE_t vxwidth,
                             sim_params
                             ):
    """
    create objects f_S and N_S that are the objects that capture
    the effect of a bulk source flowing in from the right

    inputs:
    S -- (ndarray, ndim=1) source function data, contains only data for vx < 0
    GCFL_int -- (ndarray, ndim=2) indexed [s,j], int(CFL) for stage s at vx j
    GCFL_frac -- (ndarray, ndim=2) indexed [s,j], frac(CFL) for stage s at vx j
    num _stages -- (int) sim_params['splitting']['number_of_stages']['a'],
                   total number of stages

    Nx  -- (int) total number of x gridpoints
    Nvx -- (int) total number of vx gridpoints
    sim_params -- (dict) simulation paramters

    outputs:
    f_S -- (ndarray, ndim=3), shape = (stages+1, x.N, vx.N) density array that is
           the contribution of the source particles to every [i,j] for
           a velocity dependent distribution function

    N_S -- (ndarray, ndim = 1), shape = (stages+1,) the number of source
            particles that reach the lower boundary

    see notebook s28 for details
    """
    cdef int S_Nvx = len(S)
    cdef int s, i, j
    cdef int k1, k2

    # initialize an array is indexed as (s,i,j) where s labels the stage (>= 1)
    cdef np.ndarray[DTYPE_t, ndim=3] f_S = np.zeros((num_stages + 1, Nx, Nvx))

    # initialize a 1D array indexed by s, the stage (s >= 1; s = 0 entry is empty)
    cdef np.ndarray[DTYPE_t, ndim=1] N_S = np.zeros(num_stages + 1)

    for s in range(1, num_stages + 1):
        for j in range(S_Nvx):

            NG = np.abs(GCFL_int[s,j]) + 1 # each j requires NG ghost cells
            g_crit = np.abs(GCFL_int[s,j]) - Nx # the critical ghost index g (in configuration, i = Nx + g) for a given j where g < g_crit reaches the wall, g >= g_crit maps on-grid
            print "g_crit = %g" % g_crit

            for g in range(NG): # indexes position of ghost cell: i = Nx + g
                # note these cells are not stored, but their effect has been
                # worked out, see notebook s28

                if 0 <= g < g_crit: # both proportions k1 and k2 reach boundary
                    print "reaches wall"
                    N_S[s] +=  S[j] * (1 + GCFL_frac[s,j])
                    N_S[s] += -S[j] *  GCFL_frac[s,j]

                elif g == g_crit: # edge case, ghost prepoints reach i in [-1,0]
                                  # lower boundary at i = -1/2

                    if -GCFL_frac[s,j] > 1/2.: # both k1 and k2 reach boundary
                        # physically k1 = -1/2, k2 = -1
                        N_S[s] +=  S[j] * (1 + GCFL_frac[s,j])
                        N_S[s] += -S[j] *  GCFL_frac[s,j]

                    elif -GCFL_frac[s,j] < 1/2.: # k2 reaches boundary, k1 on grid
                        # physically k1 = 0, k2 = -1/2

                        f_S[s,k1,j] += S[j] * (1 + 2 * GCFL_frac[s,j])
                        N_S[s] += -2 * S[j] *  GCFL_frac[s,j]

                elif np.max((g_crit + 1, 0)) <= g < NG-1:

                    # both proportions map on-grid to points (g + GCFL.int[s,j]) and (g + GCFL.int[s,j] - 1)
                    k1 = g + GCFL_int[s,j]
                    k2 = k1 - 1

                    f_S[s,k1,j] +=  S[j] * (1 + GCFL_frac[s,j])
                    f_S[s,k2,j] += -S[j] *  GCFL_frac[s,j]

                elif g == NG-1:
                    # only k2 proportion maps on-grid
                    # physically k2 = Nx-1, k1 = Nx (off-grid, ghost region)

                    k2 = Nx-1
                    f_S[s,k2,j] += -S[j] *  GCFL_frac[s,j]

                else:
                    # this case (g > NG-1) never happens, loop is for 0 <= g <= NG-1, included for aesthetics
                    pass

    # up to this point total number that has reached the wall has been stored in
    # a 1D vector N_S[s] for each stage, where N_S[s] = np.sum(chi[k]*S[s,j])
    # need to integrate over space:
    #
    #         $$N_S = \int dx \int dv_x S(x_g -> x[k] <= x_wall)
    #
    #               \simeq \sum_k \sum_g \sum_j \Delta x \Delta v_x chi_k S
    #
    # factor in multiplicative prefactors to all entry N_S[s]:
    N_S *= xwidth
    N_S *= vxwidth

    return f_S, N_S
```
