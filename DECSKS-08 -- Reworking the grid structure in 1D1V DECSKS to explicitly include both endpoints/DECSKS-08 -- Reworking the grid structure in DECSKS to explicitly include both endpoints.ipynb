# Reworking the grid structure in DECSKS to explicitly include both endpoints

# A. Motivation: application of more types of boundary conditions

In the first versions of DECSKS (herafter referred to as DECSKS-v1.1 to be correspondent with github versioning), the only aim was periodic boundary conditions. Because of this restricted immediate goal, the grid setup was tuned specifically for this task in an esoteric way that does not have the flexibility to straightforwardly incorporate other boundary conditions (in short, because there are <i>not</i> two boundary points in this gridding, but one endpoint that represents both). To explain the setup, we make mention there are two ways that a grid can be set up for a domain $x\in [a, b]$ where a periodic boundary condition on the state function $f$ can be enforced. One way is to treat the two endpoints as distinct, and enforce an equality in the value of $f$ at these points. The other is to estabish an equivalence between the two grid points, collapsing them into one. Thus, the fundamental difference between both grids (where the labels will be explained shortly) can be summarized as:

#### How a periodic boundary condition is enforced on each grid in an advection problem $\partial_t f + v\partial_x f = 0$:

\begin{eqnarray*}
\text{no ghost cell grid: }\qquad \qquad f(b) & \equiv & f(a),\qquad \qquad x\in [a,b) \\
&&\\
\phantom{\text{no }}\text{ghost cell grid: } \qquad \qquad f(b) & = & f(a), \qquad \qquad x\in [a,b]
\end{eqnarray*}

This appears inappropriate or otherwise unenforceable to a reader who does not have the immediate context that is being captured here. The enforcement of a periodic boundary condition here is <b>in the context of advection</b> of a measurably quantity (generically, we refer to this arbitrary quantity as a "particle", or sometimes when convenient we speak of a density packet (collection of particles over the cell volume)) that starts at one grid point and is pushed by a velocity $v = v(t,x)$ to an ending point at the conclusion of a some time step $\Delta t\in\mathbb{R}^+$. The scope of the above periodic boundary conditions then is seen to be enforced in that if in the top grid, should a particle start at some $x \neq b$ and is pushed to $x = b$, then in the top grid we actually map it to $x = a$, since these points are treated as one (hence the use of $\equiv$). In the bottom grid for the same situation, when a particle is pushed to a point $x = b$, it is sticks but with the enforcement that however many particles end up at $x = b$ is equal to that at $x = a$ (hence the use of $=$). The gridding in DECSKS-v1.1 was such that a ghost cell (GC) was implicitly involved behind-the-scenes. To illustrate what is meant further, we consider an example with a domain length $L = 1$ with $N_x = 10$ grid points where the left endpoint is $x_0 = 0$. The two grids differ as follows:

```python
from IPython.display import Image
Image(filename='./pyfiles/fig/both_grids.png')
```

DECSKS-v1.1 uses the bottom grid, which has a ghost cell that begins at $x_9 = 1 - \Delta x_{GC}$, and terminates at $x_{10} = x_9 + \Delta x_{GC} = (1 - \Delta x_{GC} + \Delta x_{GC} = 1$, where $f(x_{10}) \equiv f(x_0)$. Below, we clear up any ambiguities that may remain by making everything explicit.

# D. DECSKS updating rule for periodic boundary conditions on both grids

DECSKS is a semi-Lagrangian solver, hence it solves $\partial_t f + v \partial_x f = 0$ along its characteristics $x = x_0 - vt$, in other words, we solve the equivalent equation

$$\frac{df}{dt} = 0 \longrightarrow \int_{MC(t)} dx f(t,x) = \int_{MC_0} dx f(t_0,x)  $$

where $MC = MC(t)$ is a control volume we follow along the characteristics in the integration termed the <i>moving cell</i>. As can be seen from the integral form above, the moving cell's phase space volume $|MC_0|$ is constant over a time step, this follows from the homogeneity of what casted as an ordinary differential equation $df/dt = 0$. In general, the postpoint will not lie exactly on a gridpoint (cell center), hence we require a means of allocating the correct amount of density to contiguous cells that the MC overlaps. Assuming a constant density in an MC in the integral formulation above, we arrive directly of the statement that the appropriate fraction that is deposited in each cell is proportional to the phase space volume overlap $\zeta$ of the cell at its postpoint (note: the language prepoint and postpoint always refers to the tracked cell centers of the moving cell). The reader is directed to preliminary examination github:dsirajud/Preliminary-examination for a proofing of the remap fraction conclusion). Below, we show an advection of a density packet starting a cell centered at $i = 2$, which is advected to an $i'' = 7.25\notin\mathbb{N}$, i.e. $v\Delta t / \Delta x = 5.25$.

```python
from IPython.display import Image
Image(filename='./pyfiles/fig/advection.png')
```

Here, the dotted lines show the overlapped cell <i>boundaries</i> whose cell centers are $i = 7$ and $i = 8$ (i.e. cell boundaries are at $i\pm 1/2$ for a cell labelled as simply $i$).  the fraction $\zeta = 0.75$, hence $1 - \zeta = 0.25$. Hence, we appropriate $\zeta = 0.75$ of the density $f(t^n,x_i)$ to cell $i = 8$, and $1 - \zeta = 0.25$ to cell $i = 7$. In general, suppose the nearest grid point is some $k$ (e.g. above it can be $k = 7$), the update rule is then ($\pm$ depending on the velocity sign)

<b> Remapping rule in terms of <i>cell length overlap</i> fraction $\zeta$ ($v \geq 0$)</b>:
$${}$$
\begin{eqnarray*}
\zeta f(t^n, x_i) & \longrightarrow & f(t^{n+1},x_k) \\
&& \\
(1 - \zeta) f(t^n,x_i) & \longrightarrow & f(t^{n+1}, x_{k+ 1})
\end{eqnarray*}

```python
from IPython.display import Image
Image(filename='./pyfiles/fig/remap.png')
```

Hence, the overlap in <i>cell length</i> (for 1D, overlap area in 2D, overlap volume in 3D+) is needed. This quantity is directly obtainable, but we can save expense by discerning the relationship of $\zeta$ (fractional cell length overlap) to a quantity $\alpha$ (distance between grid points) since the latter is a quantity that can be referenced directly from storage in the code, whereas the former needs to be calculated. The remapping rule above is translated in terms of distances between the postpoint and gridpoints as:

<b> Remapping rule in terms of <i>fractional distance between postpoint and gridpoints</i> $\alpha$ ($v \geq 0$)</b>
$${}$$
\begin{eqnarray*}
(1 - \alpha) f(t^n, x_i) & \longrightarrow & f(t^{n+1},x_k) \\
&& \\
\alpha f(t^n,x_i) & \longrightarrow & f(t^{n+1}, x_{k+1})
\end{eqnarray*}

Where we show the case for the velocity $v \geq 0$, the negative velocity case follows in an obvious manner.

To implement this advection-remapping procedure for an MC at $i$, we require several quantities (we refer to a postpoint as "raw" before boundary conditions have been applied):


<b>DECSKS remapping rule "ingredients list"</b>
<ol>
<li> Compute the advected distance $v\Delta t$
<li> Compute the raw postpoint $i'' = i + \mathcal{C}$, where $\mathcal{C} = \frac{v\Delta t}{\Delta x}\in\mathbb{R}$ gives the number of cells advected (in other words, it is the Courant-Friedrichs-Lewy number).
<li> Compute a nearest neighbor grid point $k$ to the postpoint $i''$ (e.g. in the above figures, $k = 7$ or $k = 8$, the user chooses which to designate as $k$, the other neighboring grid point is $k\pm 1$ by definition). In DECSKS, we decide to always designate $k$ such that the remaining fraction of the CFL number retains the same sign as $v$, hence
$${}$$
$$\mathcal{C} := \frac{v\Delta t}{\Delta x} = S + \alpha, \quad S\in\mathbb{Z}, \alpha \in [0,1)$$

$${}$$
$$\text{nearest neighbor gridpoint to postpoint: }k = i + S, \qquad \text{where } S = \begin{cases}
\text{floor}(\mathcal{C}) & v \geq 0 \\
\\
\text{ceil}(\mathcal{C}) & v < 0\end{cases}$$
$${}$$
<li> Compute the proportion $\alpha$, the fractional distance between cell centers:
$${}$$
$$\alpha = \mathcal{C} - S$$
$${}$$

hence, $\text{sign}(\alpha ) = \text{sign}(v)$, from our designations of how to compute $S$ above. Thus, the density is remapped to $k$ (resp. $k\pm 1$) with proportion $|\alpha|$ (resp. $|1 \pm \alpha|$).
</ol>

The care that is needed here is to discern how to properly find the indices $k$ and $k \pm 1$ for periodic boundary conditions.

The reason one might use ghost cells (GC grid), such as DECSKS-v1.1 employs, is due to the fact that <b>periodic boundary conditions require only one endpoint to be updated</b>. Implementation-wise, the other endpoint is not involved as part of the active problem inasmuch as the value of the other endpoint is duplicated. Thus, a GC grid differs from an NGC grid in its exclusion of this passive point whose value is just copied at each time step.

These boundary conditions can be compactly calculated without conditional instructions by deferring to the use of modular arithmetic. That is, for a GC grid we have the DECSKS time step taking the following form

<table>
<th>Algorithm 1 : periodic boundaries on a GC grid -- $\Delta x = (b - a) / N_x, x\in \mathcal{D}_x = [a,b)$</th>
<tr><td><ol>
<li> for $i\in [0,1,2, \ldots , N_x - 1]: \, x_i\in\mathcal{D}_x$, $t\in [t^n, t^{n+1}]$, a density packet at $f(t^n,x_i)$ at a prepoint $i$ is advected to a postpoint $i' = i + \mathcal{C}$ ($\mathcal{C} = v\Delta t / \Delta x$) and reallocated to contiguous $\{k_1, k_2\}$, according to: 
$${}$$
<ol>
<li> Normalized moving cell displacement $\mathcal{C} := \frac{v\Delta t}{\Delta x} = S + \alpha$, where
$${}$$
$$S = \begin{cases}
\text{floor}(\mathcal{C}) & v \geq 0 \\
\\
\text{ceil}(\mathcal{C}) & v < 0\end{cases}$$
$${}$$
and
$${}$$
$$\alpha = \mathcal{C} - S$$
$${}$$
Note: by construction $\text{sign}(\alpha ) = \text{sign}(v)$.

<li> Determine the mapping $i \mapsto \{k_1', k_2'\}$ to contiguous cells with centroids $k_1', k_2'$, where
$${}$$
\begin{eqnarray*}
k_1' & = & i + S \\
&& \\
k_2' & = & \begin{cases} k_1' + 1 & v \geq 0 \\
& \\
k_1' - 1 &  v < 0\\ \end{cases}
\end{eqnarray*}


<li> <u>Periodic boundary conditions</u>: gridpoint $i''$ (hence $k_1', k_2$) can be outside domain, determine the final cells $k_1, k_2$ to be remapped to:
$${}$$
\begin{eqnarray*}
k_1 & = & k_1' \,(\text{mod } N_x) \\
&& \\
k_2 & = & k_2' \,(\text{mod } N_x)
\end{eqnarray*}
$${}$$
<li> update density container at time $t^{n+1}$: $f(t^{n+1},x_{i'}) +\hspace{-0.25em}= f(t^n,x_i)$
</ol>
$${}$$
repeat for all $i$</td></tr>
</ol>
</table>

The necessary changeover for the same boundary condition enforcement on an NGC (no ghost cell) grid, is to designate one of the two endpoints as "active," and the other as "passive." This is an acknowledgement that only one gridpoint needs to be updated, the other endpoint value is enforced by the equality $f(x_{\text{passive, endpoint}}) = f(x_{\text{active, endpoint}})$. It is clear that which endpoint plays what role has does not matter in the end. Here, we have the proclivity to designate the left endpoint as the active point, and permit the right endpoint to be passive. That is, we solve the problem on $x\in [a, b)$ and complete the solution by assigning $f(x_b) = f(x_a)$, i.e. we have the stepthrough $\text{DECSKS solution} + \text{periodic BC enforcement} = f(x_i), \, i\in [0,N_x-1]: \, x_i\in [a,b]$, in other words we assemble the solution per $[a,b)\cup\{b\} = [a,b]$.  Thus, the algorithm takes the form:

<table>
<th>Algorithm 2 : periodic boundaries on an NGC grid -- $\Delta x = (b - a) / (N_x - 1), x\in \mathcal{D}_x = [a,b]$</th>
<tr><td><ol>
<li> On the "active grid": for $i\in [0,1,2, \ldots , N_x - 2]): \, x_i\in\mathcal{D}_x\backslash \{b\}$, $t\in [t^n, t^{n+1}]$, a density packet at $f(t^n,x_i)$ at a prepoint $i$ is advected to a postpoint $i' = i + \mathcal{C}$ ($\mathcal{C} = v\Delta t / \Delta x$) and reallocated to contiguous $\{k_1, k_2\}$, according to: 
$${}$$
<ol>
<li> Normalized moving cell displacement $\mathcal{C} := \frac{v\Delta t}{\Delta x} = S + \alpha$, where
$${}$$
$$S = \begin{cases}
\text{floor}(\mathcal{C}) & v \geq 0 \\
\\
\text{ceil}(\mathcal{C}) & v < 0\end{cases}$$
$${}$$
and
$${}$$
$$\alpha = \mathcal{C} - S$$
$${}$$
Note: by construction $\text{sign}(\alpha ) = \text{sign}(v)$.

<li> Determine the mapping $i \mapsto \{k_1', k_2'\}$ to contiguous cells with centroids $k_1', k_2'$, where
$${}$$
\begin{eqnarray*}
k_1' & = & i + S \\
&& \\
k_2' & = & \begin{cases} k_1' + 1 & v \geq 0 \\
& \\
k_1' - 1 &  v < 0\\ 
\end{cases}
\end{eqnarray*}

<li> <u>Periodic boundary conditions 1</u>: gridpoint $i''$ (hence $k_1', k_2$) can be outside domain, determine the final cells $k_1, k_2$ to be remapped to:
$${}$$
\begin{eqnarray*}
k_1 & = & k_1' \,(\text{mod } (N_x - 1)) \\
&& \\
k_2 & = & k_2' \,(\text{mod } (N_x - 1))
\end{eqnarray*}
$${}$$
<li> update density container at time $t^{n+1}$: $f(t^{n+1},x_{i'}) +\hspace{-0.25em}= f(t^n,x_i)$
</ol>
$${}$$
repeat for all remaining points on the active grid $i:\, x_i\in\mathcal{D}_x\backslash\{b\}$

<li> <u>Periodic boundary conditions 2</u>: $f(t^{n+1}, x_{N_x - 1}) = f(t^{n+1}, x_0)$
</td></tr>
</ol>
</table>


<b>Thus, the changeover from a GC to NGC grid amounts to only 3 changes</b>:

<ol>
<li> $\Delta x := (b - a) / Nx \longrightarrow (b - a) / (N_x - 1)$
<li> loop algorithm 1 over $i = 0, 1, \ldots , N_x - 2$, the active grid.
<li> account for periodic BC in advected indices $k_1', k_2'$ with modular arithmetic on the active grid, i.e. $k_{\ell} = \text{mod}\,(k_{\ell}', N_x) \longrightarrow \text{mod}\,(k_{\ell}', N_x - 1)$, where $\ell = \{1,2\}$.
<li> after each time step is concluded, duplicate the left endpoint according to periodic BC: $f(t^{n+1}, x_{N_x - 1}) = f(t^{n+1}, x_0)$.
</ol>

DECSKS is set up so that we loop through the entire grid $i = 0, 1, \ldots N_x - 1$ ($N_x$ gridpoints) currently, but just omits the duplicate gridpoint, hence the domains do not extend between some $[a,b]$ but exist in $[a,b)$. In order to include the effect of periodic boundary conditions, we wish to adjust the spacing per (1) above, so that we encompass $[a,b]$. In this way, we still wish to retain looping over all gridpoints $i = 0, 1, 2, \ldots , N_x - 1$ in general. However, we note the changeover item (2) conflicts with this notion (only loops over $N_x - 1$ cells). It is <b>not</b> of interest to rewrite the code to loop over $N_x - 1$ gridpoints so that periodic boundary conditions can be met. This rewinds us to the same circumstance we are currently in, where the domain only is applicable to a periodic situation. A workaround to permit the enforcement of periodic BCs, while also making other boundary conditions able to be enforced on the endpoints is not to rewrite the loops, but instead to define $N_x$ based on the context. For periodic BCs, we write $N_x \mapsto N_x - 1$ to the phase space variable, for all other boundary conditions we leave $N_x = N_x$.

## DECSKS code changes

In DECSKS, the evolved <b>phase space variables are class instances</b> whose class attributes ($N_x, L_x, \Delta x, \ldots $) are directly callable in a natural way ($\text{x.N, x.L, x.width}, \ldots$) which are established in the constructor $\text{lib.domain}$. To see how the changes prescribed above pan out, we provide the constructor function for each phase space variable (DECSKS.lib.domain) for direct reference.

#### lib.domain

```python
import numpy as np

class Setup:
    """Returns (t, x, y, z, vx, vy, vz) instances

    inputs:
    sim_params -- (dict) simulation parameters from read_input()
    var -- (str) v, {x,y,z}, or t
    dim -- (str) if var = 'v', dim = x, y, z must be supplied
                 else dim = None by keyword

    outputs:
    self -- (instance) time or phase space variable
    """
    def __init__(self, sim_params, var):
        if var.lower() != 't':
            # var = 'x','y', or 'z'; dim = None
            self.N = sim_params['N' + var]
            self.a = float(sim_params['a' + var])
            self.b = float(sim_params['b' + var])
            self.L = float(self.b - self.a)
            self.width = self.L / self.N
            self.str = var
            # self.gridpoints and self.prepoints are
            # identical, but used in different loops in
            # the implementation depending on which
            # label is most suitable for a given context
            self.gridpoints = np.array(range(self.N))
            self.prepoints = np.array(range(self.N))
            self.postpoints = np.zeros(self.N)

            self.cells = self.generate_Eulerian_mesh(self.N)
            self.MCs = np.zeros(self.N)

        else:
            # var = 't', dim = None
            self.N = sim_params['N' + var]
            self.a = float(sim_params['a' + var])
            self.b = float(sim_params['b' + var])
            self.T = float(self.b - self.a)
            self.width = self.T / self.N
            self.stepnumbers = np.array(range(1,self.N+1))
            self.times = self.generate_Eulerian_mesh(self.N+1)
            self.str = var

    def generate_Eulerian_mesh(self, Num):
        """Mesh generator for domain cells, w = {x,y,z,vx,vy,vz,t}"""
        w = np.zeros(Num)
        for i in range(Num):
            w[i] = self.a + i*self.width
        return w

    def generate_Lagrangian_mesh(self, z, vz, dt):
        """Advects all MCs one time step for a phase space variable z.

        inputs:
        z -- (ndarray, dim=1) phase space variable mesh
        vvz -- (ndarray, dim=1) velocity vector whose entries pair with each MC
        dt -- (float) time step taken

        outputs:
        z_MC -- (ndarray, dim=1) array of raw positions of each MC after push
        """

        z_MC = z + vz*dt

        return z_MC
```

This constructor function draws from a dictionary $\text{sim_params}$, which holds all the information specified in the collection of input files required to run the simulation $\text{DECSKS/etc/params.dat}$. Inside $\text{params.dat}$, the number of gridpoints $\text{N}$ is specified for each phase space variable.

In lib.domain (above), the number of gridpoints is recorded $\text{self.N}$, where $\text{self} = \{x,y,z\}$, and so on for other relevant parameters (e.g. $\text{self.L}$). Thus, we for an instance $\text{self} = x$, the above defines

<blockquote><u><i>defitions pertaining to the above lib.domain</i></u></blockquote>
\begin{eqnarray*}
\text{self} & := & x \\
&& \\
\text{self.N} & := & N_x\\
&& \\
\text{self.width}  & = & \mathrm{\frac{self.L}{self.N}} := \Delta x
\end{eqnarray*}

For all boundary conditions, we require a complete grid, thus we should compute $\Delta x = L / (N_x - 1)$. In fact, a <i>so</i> many references to $\text{x.N} = N_x$ must be modified according to $N_x \rightarrow N_x - 1$ when we decide to incorporate a full grid that we are tempted to consider just re-establishing the definition of $\text{x.N} = N_x - 1$; however, we find in deeper branches of the code we still are besetted with requiring to understand a fundamental difference between the number of active cells (those that are evolved) vs. those that are entire grid (including passive gridpoints). Thus, this first idea does not pan out consistently; however, we can modify the least amount of code to generalize our stepthrough here by establishing two quantities whose roles are clearly defined. That is, we create two clear class attributes:

    z.Ngridpoints -- (int) number of total gridpoints on the domain [a,b]
    z.N -- (int) number of active grid points (those that are evolved in the solver)
    
<b>Note that in all cases but the periodic boundary condition, we have $\text{self.N} = \text{self.Ngridpoints}$</b>.

Necessarily a conditional statement must be included here, to the effect of:

    if sim_params['BC'].lower() == 'periodic':
        self.N = self.Ngridpoints - 1
    else:

```python
import numpy as np

class Setup:
    """Returns (t, x, y, z, vx, vy, vz) instances

    inputs:
    sim_params -- (dict) simulation parameters from read_input()
    var -- (str) v, {x,y,z}, or t
    dim -- (str) if var = 'v', dim = x, y, z must be supplied
                 else dim = None by keyword

    outputs:
    self -- (instance) time or phase space variable
    """
    def __init__(self, sim_params, var):
        if var.lower() != 't':
            # var = 'x','y', or 'z'; dim = None

            # set up number of gridpoints (self.Ngridpoints)
            # vs. number of active gridpoints self.N (can be equal) that are evolved
            self.Ngridpoints = sim_params['N' + var] # number of total gridpoints
            if sim_params['BC'].lower() == 'periodic':
                self.N = self.Ngridpoints - 1
            else:
                self.N = self.Ngridpoints

            self.a = float(sim_params['a' + var])
            self.b = float(sim_params['b' + var])
            self.L = float(self.b - self.a)
            self.width = self.L / (self.Ngridpoints - 1)
            self.str = var

            self.prepoints = np.array(range(self.N))
            self.postpoints = np.zeros(self.N) # container, to be filled at each timestep
            self.MCs = np.zeros(self.N) # container, to be filled at each timestep

            # for convecting cells
            self.prepointvalues = np.zeros(self.N)
            self.prepointvalues = self.generate_Eulerian_mesh(self.N)

            # for plots
            self.gridpoints = np.array(range(self.Ngridpoints))
            self.gridvalues = self.generate_Eulerian_mesh(self.Ngridpoints)

        else:
            # var = 't', dim = None
            self.N = sim_params['N' + var] # number of timesteps
            self.Ngridpoints = self.N + 1 # total number of time gridpoints
            self.a = float(sim_params['a' + var])
            self.b = float(sim_params['b' + var])
            self.T = float(self.b - self.a)
            self.width = self.T / self.N
            self.stepnumbers = np.array(range(1,self.N+1))
            self.times = self.generate_Eulerian_mesh(self.N+1)
            self.str = var

    def generate_Eulerian_mesh(self, Num):
        """Mesh generator for domain cells, w = {x,y,z,vx,vy,vz,t}"""
        w = np.zeros(Num)
        for i in range(Num):
            w[i] = self.a + i*self.width
        return w

    def generate_Lagrangian_mesh(self, z, vz, dt):
        """Advects all MCs one time step for a phase space variable z.

        inputs:
        z -- (ndarray, dim=1) phase space variable mesh
        vvz -- (ndarray, dim=1) velocity vector whose entries pair with each MC
        dt -- (float) time step taken

        outputs:
        z_MC -- (ndarray, dim=1) array of raw positions of each MC after push
        """

        z_MC = z + vz*dt

        return z_MC
```

# D. Verification: setup

We solve the advection equation:

$$\partial_t f + 1\cdot \partial_x f = 0, \qquad \qquad f = f(t,x)$$

\begin{align}
x\in\mathcal{D} & = [-\tfrac{1}{2}, \tfrac{1}{2}], \qquad t\in [0,1]  \\[0.8em]
    \text{initial condition: }\phantom{condition: } f_0(x) & = \frac{1}{2}e^{-\left(\tfrac{x + 0.2}{0.03}\right)^2} + e^{-\left(\tfrac{x}{0.06}\right)^2} + \frac{1}{2}e^{-\left(\tfrac{x - 0.02}{0.03}\right)^2}  \\[0.8em]
\text{periodic boundary condition: } f(t,x) & = f(x + n\cdot 1), \, n\in\mathbb{Z}
\end{align}

We discretize according to:

$$t^n = n\Delta t, \quad n = 0, 1, 2 , \ldots N_t\in\mathbb{N}, \quad \Delta t = \frac{1}{N_t}$$

$$x_i = 0 + i\cdot\Delta x, \quad i = 0, 1, 2, \ldots, N_x - 1\in\mathbb{N}, \quad \Delta x = \frac{0.5 - (-0.5)}{(N_x - 1)}$$

where $L = 1$ is the domain length, hence we have completed the solution over the entire real line by periodic extension as any abscissa $x + n\cdot 1$ is equal to the state function $f$ inside the domain $x\in\mathcal{D}$. The velocity is chosen to be $v = 1.0$, so that in $v(n\Delta t) = 1.0 = L$ a full period is completed over $n = \frac{1.0}{v\Delta t} = \frac{1.0}{1.0\cdot\Delta t} = \frac{1}{\Delta t} = \frac{N_t}{1.0} = N_t$, for the given time duration $t\in [0,1]$.
