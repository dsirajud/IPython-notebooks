# A. Introduction

The collisionless matrix sheath problem develops from the electrostatics influence on local plasma species populations in the vicinity of a sink (boundary). We regard the formation of a positive space charge within several Debye lengths of the boundary ("edge", $n_i > n_e$). A fluid analysis produces the criterion that the ion velocity entering the "sheath" are accelerated to the sound speed ($v_{i,se} = c_i$). By considering the dynamics from the bulk plasma looking towards the edge ($\ell_{\nabla} \gg 1$), and from the edge looking outboard ($\ell_{\nabla} \ll 1$), the solutions for both scales are matched per usual strategem requiring not an inequality, but an equality. In practice, the rough agreement that ions enter the sheath supersonically is commonly observed within margin. The criterion has gained momentum over the years to be accepted as a rule-of-thumb regarding the nonlinear sheath physics. More dangerously, the Bohm criterion has become accepted to the point where shortcuts in simulations are afforded by including it as a "boundary condition" when simulating the edge. The convenience is clear, requiring supersonic velocities of ions at the sheath boundary of a simulation domain saves significant computational expense of the complicated presheath physics in order to (what is hoped) to arrive at the same result (the Bohm criterion). The sticking issue that this practice is often done yet it is acknowledged that the Bohm criterion appears to be less tenable than it was at its inception. It is necesary to create a kinetic sheath in simulation order to see how well this criterion is upheld.

## A.1 Fluid derivation (classic Bohm criterion)

The classic Bohm criterion follows from the moment-based (fluid) approach, we consider the collisionless boundary problem modelling the electrons in the (adiabatic) limit. Begin with the kinetic (Vlasov) equation:

$$\partial_t f_e + \vec{v}\cdot \vec{\nabla}_xf_e  - \tfrac{e\vec{E}}{m_e}\cdot\vec{\nabla}_vf_e = 0$$

Taking the first moment ($\int d^3v\,\vec{v}$) we identify the required averages $n_e \vec{V}:= \int d^3v \vec{v} f_e$, where $\vec{V}$ is the average velocity. The third term can be written as:

$$\int d^3v \vec{v} \frac{e\vec{E}}{m_e}\cdot\vec{\nabla}_vf_e = \frac{e\vec{E}}{m_e}\cdot\int d^3v \vec{v} \vec{\nabla}_vf =  \frac{e\vec{E}}{m_e}\cdot\int d^3 v \left[\vec{\nabla}_v(\vec{v} f_e) - (\vec{\nabla}_v \vec{v})f_e\right]$$

The first term on the right-hand side $\int_{V} d^3 v \vec{\nabla}_v\cdot (\vec{v}f_e) = \int_{\partial S} d^2\vec{v}\cdot\vec{v}f_e \rightarrow 0$ at $|v| \rightarrow \infty$. The second term is obvious, note that:

$$\vec{\nabla}_v \vec{v} = \partial_{v_i} v_j \hat{x}_i\hat{x}_j \delta_{ij} = \hat{x}_i\hat{x}_i =\vec{\vec{I}}$$

so the above becomes:

$$\int d^3v \vec{v} \frac{e\vec{E}}{m_e}\cdot\vec{\nabla}_vf_e = -\frac{e\vec{E}}{m_e}\cdot\int d^3 v \vec{\vec{I}} f_e = -\frac{e\vec{E}}{m_e}\int d^3 v f_e = -\frac{e\vec{E}}{m_e} n_e$$

And, the middle term:

$$\int d^3 v \vec{v} \vec{v}\cdot \vec{\nabla}_xf_e = \int d^3 v [\vec{\nabla}_x\cdot (\vec{v}\vec{v}f_e) - f_e\vec{\nabla}_x \cdot (\vec{v}\vec{v})] = \vec{\nabla}_x\cdot \int d^3 v \vec{v}\vec{v}f_e$$

Since the final term $\vec{\nabla}_x \cdot (\vec{v}\vec{v}) = \partial_{x_i} \hat{x_i}\cdot v_jv_k\hat{x_j}\hat{x_k} = 0 \hat{x}_i\hat{x}_j$, since $\partial_{x_{\ell}} v_m = 0$ for all $\ell , m$.

It is convenient to rewrite the dyad in terms of the mean velocity $\vec{V} = \int d^3 v \vec{v}f_e$, so that physical quantities (pressures) can be readily identified. For example:

$$\vec{v}\vec{v} = (\vec{v} - \vec{V})(\vec{v} - \vec{V}) - \vec{V}\vec{V} + \vec{v}\vec{V} + \vec{V}\vec{v}$$

Then, the above is seen to be:

\begin{eqnarray*}
\int d^3 v \vec{v} \vec{v}\cdot \vec{\nabla}_xf_e & =  &\vec{\nabla}_x\cdot \int d^3 v \vec{v}\vec{v}f_e \\
&= &  \vec{\nabla}_x\cdot \int d^3 v [(\vec{v} - \vec{V})(\vec{v} - \vec{V}) - \vec{V}\vec{V} + \vec{v}\vec{V} + \vec{V}\vec{v}]f_e \\
&= & \vec{\nabla}_x\cdot \int d^3 v (\vec{v} - \vec{V})(\vec{v} - \vec{V})f_e - \vec{\nabla}_x\cdot \int d^3 v\vec{V}\vec{V}f_e + \vec{\nabla}_x\cdot \int d^3 v\vec{v}\vec{V}f_e + \vec{\nabla}_x\cdot \int d^3 v\vec{V}\vec{v}f_e \\
&= & \vec{\nabla}_x\cdot \int d^3 v (\vec{v} - \vec{V})(\vec{v} - \vec{V})f_e - \vec{\nabla}_x\cdot \vec{V}\vec{V}n_e + \vec{\nabla}_x\cdot\vec{V}\vec{V}n_e + \vec{\nabla}_x\cdot\vec{V}\vec{V}n_e \\
&= & \vec{\nabla}_x\cdot \int d^3 v (\vec{v} - \vec{V})(\vec{v} - \vec{V})f_e + (\vec{\nabla}_x\cdot\vec{V}\vec{V}n_e)\\
\int d^3 v \vec{v} \vec{v}\cdot \vec{\nabla}_xf_e & = & \frac{1}{m_e}\vec{\nabla}_x\cdot \vec{\vec{P}} + \vec{\nabla}_x\cdot(\vec{V}\vec{V}n_e)
\end{eqnarray*}

So the momentum equation is:

$$\frac{\partial(n_e\vec{V})}{\partial t} + \vec{\nabla}_x\cdot(n_e\vec{V}\vec{V}) + \frac{1}{m_e}\vec{\nabla}_x\cdot \vec{\vec{P}} + \frac{e\vec{E}}{m_e} n_e = 0$$

Where we have the pressure tensor $\vec{\vec{P}}$. Often, this equation is arranged with intertial terms on the left:

$$\frac{\partial(n_e\vec{V})}{\partial t} + \vec{\nabla}_x\cdot(n_e\vec{V}\vec{V}) = -\frac{1}{m_e}\vec{\nabla}_x\cdot \vec{\vec{P}} - \frac{e\vec{E}}{m_e} n_e$$

At this point, applying the derivatives on the left-hand sides causes the continuity equation to unfold naturally, which allows the removal of some terms given continuity implies both terms sum to zero. This achieves the form most frequently quoted as the <b>momentum equation</b>:

$$n_e\frac{\partial\vec{V}}{\partial t} + n_e \vec{V}\vec{\nabla}_x\cdot \vec{V} = -\frac{1}{m_e}\vec{\nabla}_x\cdot \vec{\vec{P}} - \frac{e\vec{E}}{m_e} n_e \qquad \underline{\text{momentum equation for electrons}}$$

Further, a partition of the pressure tensor is also convenient and clarifying:

$$\vec{\vec{P}} = p\vec{\vec{I}} + \vec{\vec{\Pi}}$$

where the scalar pressure $p = \frac{1}{3}\text{Tr}(\vec{\vec{P}})$.  Note, the pressure tensor itself is a second rank symmetric tensor, thus the trace is invariant under orthonormal coordinate transformations, hence a scalar pressure of the same value independent of coordinate system can always be defined the same way. We often consider simpler circumstances, for example if the pressure is isotropic, then the viscosity tensor $\vec{\vec{\Pi}}= \vec{\vec{0}}$. So, we reduce the pressure tensor in isotropic cases to be the following and its divergence accordingly becomes:

$$\vec{\vec{P}} = p\vec{\vec{I}} \Rightarrow \vec{\nabla}_x\cdot \vec{\vec{P}} = \vec{\nabla}_x\cdot p\vec{\vec{I}} = \partial_{x_i}\hat{x}_i\cdot p\hat{x}_j\hat{x_k} = \partial_{x_i} p \hat{x}_i = \vec{\nabla}_xp$$

Thus,

$$\frac{\partial(m_en_e\vec{V})}{\partial t} + \vec{\nabla}_x\cdot(m_en_e\vec{V}\vec{V}) = -\vec{\nabla}_x p - e\vec{E} n_e$$

Adiabatic means slowly varying, so that the inertial terms on the left can be discarded and we consider the steady state of the electron species:

$$0 = -\vec{\nabla}_x p -  e\vec{E}n_e$$

We have tacitly (yet overtly) asserted we are not considering magnetic fields. Thus, in electrostatics the usual electric potential is afforded by virtue of the form of Faraday's law: $\vec{E} = -\vec{\nabla}_x \phi$. An equation of state can also be asserted for our sufficiently regular circumstances: $p = n_eT_e$. In general, this fluid parameter (an average) is a function of the configuration variable $x$, here we are considering a limited case where the electrons respond only to electrostatic interactions. The electrons being sufficiently mobile respond rapidly to any local variations that develop and act quickly to regularize any charge deviations as best they can (effectively accomplished in the bulk [quasineutrality], partially accomplished in the thin edge region). Notwithstanding, the temperature gradient scale length is much larger than edge characteristic scale lengths. Hence an isothermal enforcement is appropriate, and not inaccurate in the thin edge. Hence, 

\begin{eqnarray*}
\vec{\nabla}_x (n_eT_e) - en_e\vec{\nabla}_x\phi & = & 0 \\
n_e T_e\vec{\nabla}_x\ln n_e - en_e\vec{\nabla}_x\phi & = & 0 \\
n_e \vec{\nabla}_x\left(T_e\ln n_e  - e\phi\right) & = & 0 \\
\vec{\nabla}_x\left(T_e\ln n_e  - e\phi\right) & = & 0
\end{eqnarray*}

So,

$$T_e\ln n_e  - e\phi = \text{constant} \qquad \text{or} \qquad \ln n_e = \frac{\text{constant}}{T_e} + \frac{e\phi}{T_e} \Rightarrow \boxed{n_e = n_{e0}\exp\left(\frac{e\phi}{T_e}\right)} \qquad \underline{\text{Adiabatic electrons}}$$

To find the steady state ($\partial_t\rightarrow 0$) distribution in the edge region, <b>consider a one-dimensional problem in $x$</b>. The first moment gives the continuity equation:

$$\partial_x (n_i v_i) = 0 \Rightarrow n_i = \frac{n_0 v_{i0}}{v_i}$$

Combining this with the adiabatic electron relationship into Poisson's equation:

$$\partial_x^2\phi = \frac{e}{\epsilon_0}(n_e - n_i) = \frac{en_0}{\epsilon_0}\left( e^{e\phi/T_e} - \frac{v_i}{v_{i0}}\right) \qquad (1) $$

To develop an expression for $v_i / v_{i0}$, consider the isobaric momentum equation in steady state for <i>ions</i> which is equivalent to:

$$\frac{1}{2}n_i\partial_x(v_i^2) - \frac{eE}{m_i} n_i = 0$$

or

$$n_i\frac{1}{2}\partial_x(v_i^2) = \frac{eE}{m_i} n_i = -n_i\frac{e}{m_i}\partial_x \phi$$

Which has a first integral that gives the velocity of ions:

$$v_i = \sqrt{v_{i0}^2 - \frac{2 e\phi}{m_iv_{i0}^2}} \Rightarrow \frac{v_i}{v_{i0}} = \sqrt{1 - \frac{2 e\phi}{m_iv_{i0}^2}}$$

Here, we have tacitly decided the reference potential at the sheath edge $\phi_{se} = 0$ for algebraic convenience, an inclusion of nonzero $\phi_se \neq 0$ only changes the result by replacing $\phi \rightarrow \phi - \phi_{se}$. Hence, while we lose generality in this decision, it is obvious how to incorporate it back in.

Recall that the sound speed $c_s^2 = (T_i + T_e) / m_i \simeq T_e / m_i$ for our collisionless situation since we expect nothing but $T_e \gg T_i$ (i.e. the electrons gain energy more easily than the ions from the field given their high mobility, so that $T_e \sim \frac{m_i}{m_e}T_i$ as usual, where $m_i / m_e \sim 1836$. Define the Mach number as the multiple that the peculiar ion speed  at the boundary condition (at some $x = x_{se}$, where $se$ denotes a sheath edge, i.e. the transition from the bulk plasma to the edge regions) is larger than the sound speed.

Then, we can write:

$$M_0^2 := \frac{v_{i0}}{c_s}^2 =  \frac{v_{i0}^2}{T_e / m_i} \Rightarrow v_{i0}^2 = \frac{M_0^2}{T_e / m_i}$$

Combining this with the equation for the ion velocity $v_i$, eq. (1) (Poisson's equation) becomes:

$$\partial_x^2\phi = \frac{e}{\epsilon_0}(n_e - n_i) = \frac{en_0}{\epsilon_0}\left( e^{e\phi/T_e} - \sqrt{1 - \frac{2 e\phi}{T_e M_0^2}} \right) $$

Finally, consider regimes where $e\phi / T_e \ll 1$, which permits expansion:

$$\partial_x^2\phi \simeq \frac{en_0}{\epsilon_0}\left( 1 + \frac{e\phi}{T_e} - \left( 1 + \frac{1}{2} \frac{2e\phi}{T_e}\frac{1}{M_0^2}\right) \right) = \frac{e^2n_0}{\epsilon_0T_e}\phi\left(1 - \frac{1}{M_0^2}\right) + O\left(e\phi / T_e  \right)^2 = \frac{1}{\lambda_D^2}\left(1 - \frac{1}{M_0^2}\right)\phi$$

Then, we have

$$\partial_x^2\phi = \frac{\phi}{\lambda_D^2}\left(1 - \frac{1}{M_0^2}\right)$$

One of the most controversial parts of this derivation is that Bohm falls back on purely mathematical concerns to finalize the result. For non-oscillatory solutions, we require

$$1 - \frac{1}{M_0^2} \geq 0 \Rightarrow 1 \geq \frac{1}{M_0^2} \Rightarrow M_0 \geq 1$$

Recalling $M_0 := v_{i0} / c_s$, this demands

$$v_{i0} \geq c_s$$

Repeating the stepthrough from the bulk plasma looking towards the edge, we arrive at the opposite restriction $v_{i0} \leq c_s$, thus the only compatible value for the entering velocity of the ions at the sheath edge is the equality:

$$\boxed{v_{i0} = c_s} \qquad \underline{\text{Bohm criterion}}$$

Many times, the condition $v_{i,0}\geq c_s$ is quoted as the Bohm criterion, especially since in actual experients there is a spreading of velocities entering the sheath given distributions of velocities exist rather than a single fluid velocity. This also broadens the transition region and obscures how to define a clean "sheath edge" for the plasma. It is clear that in real plasmas, if we track the point of immediate departure from quasineutrality $n_i \simeq n_e$, this point tends to always correspond significantly inboard from the wall, much further than one would normally interpret in a $n$ vs. $x$ plot as what appears to be the sheath edge. Different conventions can be adopted that are more observationally consistent, e.g. the $x$ coordinate where we have a deviation by 1%, or perhaps 10% can be considered. This is done often in literature.

#### after note:

When is $e\phi / T_e \ll 1$ representative of the plasma? Note that we can write this in terms of the Debye length $\lambda_D^2 = \epsilon_0 T_e / (n_0 e^2)$ 

$$\frac{e\phi}{T_e} = \frac{e\epsilon_0}{\lambda_D^2 n_0} \phi$$

In typical plasmas we have $\lambda_D \sim 10^{-5} - 10^{-4}$ m, the constant $\epsilon_0 \sim 10^{-12} \mathrm{m^{-3} kg^{-1} s^4 A^2}$, so it is not hard to have this condition be met. Only in extremely rareified plasmas (e.g. astronomical plasmas with length scales on correspondingly extreme lengths by everyday experiment standards) might we find this not met in anything considered a plasma. To be even more blatant about it, note that we can write this as:

$$\frac{e\phi}{T_e} = \frac{e\epsilon_0}{\lambda_D^2 n_0^{2/3}n_0^{1/3}} \phi =\frac{e\epsilon_0}{(\lambda_D^3 n_0)^{2/3}n_0^{1/3}} \phi$$

Which states this quantity is small ($\frac{e\phi}{T_e} \ll 1$) so long as $n_0\lambda_D^3 \gg 1$. That is, that there exists many particles in a Debye cube. This is part of the definition of a plasma.



# A.2 Kinetic derivation of generalized Bohm criterion

The kinetic criterion is obtained even more quickly than the fluid stepthrough. We aim to find a criterion that arises in the steady state system, just as before.

Begin with the Vlasov equation for a species $\alpha$ at steady state ($\partial_t \rightarrow 0$):

$$\vec{v}\cdot \vec{\nabla}_x f_{\alpha} + \frac{q_{\alpha}}{m_{\alpha}} \vec{E}\cdot \vec{\nabla}_v f_{\alpha} = 0$$ 

Considering any component $x_i = \{x_1, x_2, x_3\}$:

$$v_{x_i}\partial_{x_i} f_{\alpha} + \frac{q_{\alpha}}{m_{\alpha}} E_{x_i}\partial_{v_{x_i}} f_{\alpha} = 0$$ 

or 

$$\partial_{x_i} f_{\alpha} + \frac{q_{\alpha}}{m_{\alpha}} E_{x_i}\frac{1}{v_{x_i}}\partial_{v_{x_i}} f_{\alpha} = 0$$ 

integrating ($\int d^3 v$) and identifying $n_{\alpha} := \int d^3 v f_{\alpha}$

$$\partial_{x_i} n_{\alpha} + \frac{q_{\alpha}}{m_{\alpha}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}}\partial_{v_{x_i}} f_{\alpha} = 0 \qquad \qquad (1)$$ 

then,

$$\partial_{x_i} n_{\alpha} = -\frac{q_{\alpha}}{m_{\alpha}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}}\partial_{v_{x_i}} f_{\alpha} = 0$$ 


In terms of species $\alpha = e,i$:

$$\partial_{x_i} n_{i} = -\frac{q_{i}}{m_{i}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}}\partial_{v_{x_i}} f_{i}$$ 
$$\partial_{x_i} n_{e} = -\frac{q_{e}}{m_{e}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}}\partial_{v_{x_i}} f_{e}$$ 

Often, the ion equation has the last integral rewritten according to the following. Note that:

$$\frac{1}{v_{x_i}} \partial_{v_{x_i}} f_i = \partial_{v_{x_i}} \left(\frac{1}{v_{x_i}}f_i\right) + \frac{1}{v_{x_i}^2}f_i$$

so that the first integral on the right-hand side becomes

$$\int d^3v\, \partial_{v_{x_i}} \left(\frac{1}{v_{x_i}}f_i\right) = \int d^2v\left(\frac{1}{v_{x_i}}f_i\right) = 0 \qquad \text{at the
surface } |v|\rightarrow 0$$

leaving:

$$-\frac{q_{i}}{m_{i}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}}\partial_{v_{x_i}} f_{i} = -\frac{q_{i}}{m_{i}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}^2} f_{i}$$

and the first moment of the steady state ion Vlasov equation is:

$$\partial_{x_i} n_{i} = -\frac{q_{i}}{m_{i}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}^2} f_{i}$$ 

Inside the sheath we have $n_i > n_e$. In the bulk plasma, we have $n_i \simeq n_e$. The sheath edge marks the start of this deviation, i.e. at the sheath edge and inside we must have

$$n_i \geq n_e \quad \Rightarrow \quad \partial_{x_i} n_i \geq \partial_{x_i} n_e$$

Inserting the above two equations, and considering the case of singly charged ions $(q_i = e, q_e = -e$) we have (sign error has occured, the correct result is shown below):

$$\frac{e}{m_{i}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}^2} f_{i} \leq -\frac{e}{m_{e}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}}\partial_{v_{x_i}} f_{e}$$ 

or,

$$\boxed{-\frac{1}{m_{e}}\int d^3 v\, \frac{1}{v_{x_i}}\partial_{v_{x_i}} f_{e} \geq \frac{1}{m_{i}}\int d^3 v\, \frac{1}{v_{x_i}^2} f_{i}} \qquad \text{Kinetic Bohm criterion}$$ 

### Recovering the fluid Bohm criterion

To recover the fluid approximation, we consider the "kinetic" fluid case, i.e. all ions move at the fluid velocity $V_i$, so that $f_i(v) = n_0\delta (\vec{v} - \vec{V}_i)$, also in a fluid description we are at local thermodynamic equilbrium, so that $f_e(v) = n_0 (m / 2\pi T_e)\exp (-m_ev^2 / (2 T_e))$ and $\int d^3 v f_e(v) = n_0$.

The criterion then takes the form (where we take $v_{x_i} \equiv v)$

\begin{eqnarray*}
-\frac{1}{m_{e}}\int d^3 v\, \frac{1}{v}\partial_{v} f_{e} & \geq & \frac{1}{m_{i}}\int d^3 v\, \frac{1}{v^2} f_{i} \\
-\frac{1}{m_{e}}\int d^3 v\, \frac{1}{v}\partial_{v} \left[ n_0\left(\frac{m_e}{2\pi T_e}\right)^{3/2}\exp \left(-\frac{m_ev^2}{2 T_e}\right)\right] & \geq &\frac{1}{m_{i}}\int d^3 v\, \frac{n_0}{v^2} \delta (\vec{v} - \vec{V}_i)  \\
-\frac{1}{m_{e}}\int d^3 v\, \frac{1}{v}\left(-\frac{m_ev}{T_e}\right) n_0\left(\frac{m_e}{2\pi T_e}\right)^{3/2}\exp \left(-\frac{m_ev^2}{2 T_e}\right) & \geq & \frac{n_0}{m_{i}}\frac{1}{V_i^2} \\
+\left(\frac{m_e}{T_e}\right)\frac{1}{m_{e}}\int d^3 v\, f_e(v) & \geq &\frac{n_0}{m_{i}}\frac{1}{V_i^2} \\
\left(\frac{m_e}{T_e}\right)\frac{1}{m_{e}}n_0 & \geq &\frac{n_0}{m_{i}}\frac{1}{V_i^2}\\
\frac{1}{T_e} & \geq &\frac{1}{m_{i}}\frac{1}{V_i^2} \\
\frac{1}{T_e} & \geq &\frac{1}{m_{i}}\frac{1}{V_i^2} \\
\frac{m_i}{T_e} & \geq &\frac{1}{V_i^2} \\
\frac{1}{c_s^2} & \geq &\frac{1}{V_i^2}
\end{eqnarray*}

Or,

$$\boxed{V_i \geq c_s}$$

So, kinetic criterion checks out in this way and seems to provide appropriate generalization.

## A.3 Note about controversy


#### on the original Bohm criterion

There is significant discussion about the claimed interpretation of the Bohm criterion as well as the tacit conditions that must be made in order to obtain his result. Bohm acknowleges this in his principal paper, though does not put the words much specificity about what is off about both the stepthrough and the resulting interpretation. Actually, it is not until the final sentence that he puts forth the interpretation widely used (read: abused) in plasma physics explanations. Bohm claims the criterion, though obtained through uncareful means seems to hold in most experimental situations within "rough[ly] 20%", i.e. that the ions do enter the sheath within the sound velocity. Thus, it is expected that a guideline like this criterion not be considered too critically, especially given it is difficult to determine where exactly in the plasma bulk-edge transition region the criterion should be met.

His final sentence reads: "Thus, it may be said roughly that a sheath begins when $V_0 = kT_e /2e$ [$V$ is the potential], remembering that this means that beyond this point the potential begins to climb rapidly and the electron density to drop very rapidly.‚Äù There are no observations that deny that this does happen with the species populations, it is the definition of a sheath edge and the climbing of the potential that does not actually follow from what Bohm obtained. To that end, this author agrees with several others in that Bohm misinterpreted his result. His solution corresponds to an asymptotic solution, whose level curve at $v_i = c_s$ (sound speed) in the natural Hamiltonian phase space ($E$, $\phi$) corresponds to an unstable manifold. Indeed, Bohm obtains a solution corresponding to the sheat being located in the limit $x\rightarrow -\infty$. He misinterprets the matching point of the two-scale analysis (bulk plasma vs. edge plasma) as being identical to the point where $V = V_0$, a reference point where $E = 0$. This isn't technically true, and can be shown straightforwardly that it deviates from this point in phase space ($E$, $\phi$) in general [2]

#### on the kinetic Bohm criterion [cf. Hegna, Callen back-and-forth in IEEE]

As shown above in section A.2, a necessary step for the kinetic criterion to be obtained is the division by a velocity parallel to the electric field (above, $v_{x_i}$). The singularities that persist have no means to be removed, so this requires that the distribution function $f_i(v_{x_i} = 0) = 0$, otherwise the right-hand side diverges at the origin in phase space, and the left-hand side integrates to a quantity proportional to the temperature normalized by a density, which clearly cannot be zero. A second thought rationalization is thet this then requires in the sheath region that there exist only one-sided velocities on the ion distribution. This is reassuring given it is the expectation in this region. If any ion escapes (threshold at $f_i(v_{x_i} \leq 0$), then this is not the sheath region. However, it is not satisfying that even basic distributions such as Maxwellian ions cannot be captured by this criterion. There is other issue that some authors contest with this rationalization as well (e.g. the kinetic description takes a continuum limit so that $\lambda_D / L \rightarrow 0$ so the sheath is infinitely thin), but as concerns the present purposes, it is sufficient to close the controversy survey here given all the main points have been outlined. One bottom line is that the kinetic criterion because of its divergenceless requirement makes the criterion experimentally untestable. The nonlinear Vlasov solver DECSKS-2.0 will provide a check on this criterion, we will see where this condition is of help, thus locating the "sheath edge." 

# B. Normalization of species Vlasov equations

## B.1.A For an evolutino equation for any species $\alpha$

In standard units, we have for a species $\alpha$ the 1D1V Vlasov equation:

$$\partial_t f_{\alpha} + v\partial_x f_{\alpha} + \frac{q_{\alpha}E}{m_{\alpha}} \partial_v f_{\alpha} = 0$$

We elect to normalize to natural units, i.e. multiples of characteristic parameters of a plasma species. Define normalized (tilde) quantities according to:

\begin{eqnarray*}
t = \frac{1}{\omega_{p\alpha}}\tilde{t} & \longrightarrow & \partial_t = \omega_{p\alpha}\partial_{\tilde{t}} \qquad \omega_{p\alpha} = \sqrt{\frac{n_{\alpha} q_{\alpha}^2}{\epsilon_0m_{\alpha}}} \\
x = \lambda_D \tilde{x} & \longrightarrow & \partial_x = \frac{1}{\lambda_D}\partial_{\tilde{x}} \qquad \lambda_D = \sqrt{\frac{\epsilon_0 kT_{\alpha}}{n_{\alpha}q_{\alpha}}} \\
v = v_{T\alpha} \tilde{v} & \longrightarrow & \partial_v = \frac{1}{v_{T\alpha}}\partial_{\tilde{v}} \qquad v_{T\alpha} = \sqrt{\frac{kT_{\alpha}}{m_{\alpha}}}\\
\end{eqnarray*}

Subsequent normalizations may be needed, and will be decided if it is seen to be appropriate in the following. Substituting all of the above into the species Vlasov equation:

\begin{eqnarray*}
\partial_t f_{\alpha} + v\partial_x f_{\alpha} + \frac{q_{\alpha}E}{m_{\alpha}} \partial_v f_{\alpha} & = & 0 \\
\left(\omega_{\alpha}\partial_{\tilde{t}}\right) f_{\alpha} + \left(\tilde{v}v_{T\alpha}\right) \left(\frac{1}{\lambda_D}\partial_{\tilde{x}}\right)f_{\alpha} + \frac{q_{\alpha}E}{m_{\alpha}}\left(\frac{1}{v_{T\alpha}}\partial_{\tilde{v}}\right) f_{\alpha} = 0 \\
\underbrace{\lambda_D\omega_{\alpha}}_{=\, v_{T\alpha}}\partial_{\tilde{t}} f_{\alpha} + \tilde{v}v_{T\alpha}\partial_{\tilde{x}}f_{\alpha} + \frac{q_{\alpha}E}{m_{\alpha}}\frac{\lambda_D}{v_{T\alpha}}\partial_{\tilde{v}} f_{\alpha} = 0 \\
v_{T\alpha}\partial_{\tilde{t}} f_{\alpha} + \tilde{v}v_{T\alpha}\partial_{\tilde{x}}f_{\alpha} + \frac{q_{\alpha}E}{m_{\alpha}}\frac{\lambda_D}{v_{T\alpha}}\partial_{\tilde{v}} f_{\alpha} = 0 \\
\partial_{\tilde{t}} f_{\alpha} + \tilde{v}\partial_{\tilde{x}}f_{\alpha} + \left(\frac{q_{\alpha}\lambda_D}{m_{\alpha}v_{T\alpha}^2}\right) E \partial_{\tilde{v}} f_{\alpha} = 0 \\
\end{eqnarray*}

Hence, as anticipated we do require one more normalized quantity. It is natural to normalize the electric field here as:

$$\tilde{E} = \left(\frac{|q_{\alpha}|\lambda_D}{m_{\alpha}v_{T\alpha}^2}\right) E = \frac{E}{\bar{E}}$$

where $\bar{E} = |q_{\alpha}|\lambda_D / (m_{\alpha}v_{T\alpha}^2) = \mathrm{[force / charge]}$ has the same units as the electric field. Note, the silent decision to normalize by the magnitude is obvious but has the effect that the charge remaining in the equation is of unit charge, i.e. the above version of the Vlasov equation then is rewritten in terms of this as:


$$\partial_{\tilde{t}} f_{\alpha} + \tilde{v}\partial_{\tilde{x}}f_{\alpha} + \frac{q_{\alpha}}{|q_{\alpha}|}\underbrace{\left(\frac{|q_{\alpha}|\lambda_D}{m_{\alpha}v_{T\alpha}^2}\right) E}_{=\, \bar{E}E \equiv \tilde{E}} \partial_{\tilde{v}} f_{\alpha} = 0$$

$$\partial_{\tilde{t}} f_{\alpha} + \tilde{v}\partial_{\tilde{x}}f_{\alpha} + \frac{q_{\alpha}}{|q_{\alpha}|}\tilde{E} \partial_{\tilde{v}} f_{\alpha} = 0 $$


or, relabelling $\tilde{w} \rightarrow w$ for $w = \{t,x,v,E\}$, we understand the equation

$$\partial_t f_{\alpha} + v\partial_xf_{\alpha} +  \frac{q_{\alpha}}{|q_{\alpha}|}E \partial_v f_{\alpha} = 0 $$

measures time in multiples of plasma periods ($\sim \omega_{p\alpha}^{-1}$) for the charged species $\alpha$, $x$ is a measure of distance as multiples of Debye lengths $\lambda_D$ for species $\alpha$, velocity is measured as multiples of the species thermal velocity $v_{T\alpha}$, and the electric field is in multiples of $m_{\alpha}v_{T\alpha}^2 / (\lambda_D q_{\alpha})$, which can be interpreted as the force field per unit charge of species $\alpha$ averaged over one Debye length at its average speed.


## B.1.B Normalization of the electron and (singly charged) ion Vlasov equations

The equations evolve alongside each other. We require a reference equation that indicates the measurement of common $t, x, v, E$ variables, whereafter we can insist on the same measurements for the second equation that ultimatley will have small differences given intertial differences ($m_{\alpha}$ and $q_{\alpha}$).

We need to choose to measure on time scales and lengths according to the shortest physical scales between $\alpha = e,i$, thus it is the electron evolution that decides our normalization. Accorindgly, from the above we permit $\alpha = e$, noting $q_e = -1$and write down the normalized electron Vlasov equation:

$$\boxed{\partial_t f_{e} + v\partial_xf_{e} - E \partial_v f_{e} = 0} \qquad \underline{\text{Normalized electron Vlasov equation}} $$

To find the normalization required for the second equation, we take a few steps backwards and start wtih the original equation for clarity. The ion equation (single charged) takes the form:

$$\partial_t f_{i} + v\partial_x f_{i} + \frac{q_iE}{m_i} \partial_v f_i = 0$$

we use consistent normalizations as with the electron equation:

\begin{eqnarray*}
t = \frac{1}{\omega_{pe}}\tilde{t} & \longrightarrow & \partial_t = \omega_{pe}\partial_{\tilde{t}} \qquad \omega_{pe} = \sqrt{\frac{n_{e} q_{e}^2}{\epsilon_0m_{e}}} \\
x = \lambda_D \tilde{x} & \longrightarrow & \partial_x = \frac{1}{\lambda_D}\partial_{\tilde{x}} \qquad \lambda_D = \sqrt{\frac{\epsilon_0 kT_{e}}{n_{e}q_{e}}} \\
v = v_{Te} \tilde{v} & \longrightarrow & \partial_v = \frac{1}{v_{Te}}\partial_{\tilde{v}} \qquad v_{Te} = \sqrt{\frac{kT_{e}}{m_{e}}}\\
\end{eqnarray*}

\begin{eqnarray*}
\partial_t f_{i} + v\partial_x f_{i} + \frac{q_{i}E}{m_{i}} \partial_v f_{i} & = & 0 \\
\left(\omega_{pe}\partial_{\tilde{t}}\right) f_{i} + \left(\tilde{v}v_{Te}\right) \left(\frac{1}{\lambda_D}\partial_{\tilde{x}}\right)f_{i} + \frac{q_{i}E}{m_{i}}\left(\frac{1}{v_{Te}}\partial_{\tilde{v}}\right) f_i = 0 \\
\underbrace{\lambda_D\omega_{pe}}_{=\, v_{Te}}\partial_{\tilde{t}} f_i + \tilde{v}v_{Te}\partial_{\tilde{x}}f_i + \frac{q_{i}E}{m_{i}}\frac{\lambda_D}{v_{T\alpha}}\partial_{\tilde{v}} f_i = 0 \\
v_{Te}\partial_{\tilde{t}} f_i + \tilde{v}v_{Te}\partial_{\tilde{x}}f_i + \frac{q_{i}E}{m_{i}}\frac{\lambda_D}{v_{Te}}\partial_{\tilde{v}} f_i = 0 \\
\partial_{\tilde{t}} f_i + \tilde{v}\partial_{\tilde{x}}f_i + \left(\frac{q_{i}\lambda_D}{m_{i}v_{Te}^2}\right) E \partial_{\tilde{v}} f_i = 0 \\
\partial_{\tilde{t}} f_i + \tilde{v}\partial_{\tilde{x}}f_i + \frac{q_i}{|q_i|}\left(\frac{|q_{i}|\lambda_D}{m_{i}v_{Te}^2}\right) E \partial_{\tilde{v}} f_i = 0 
\end{eqnarray*}

Recall our normalization for the electric field are characterized not by ion quantities, but by the electron inertia:

$$\tilde{E} = \left(\frac{|q_{e}|\lambda_D}{m_{e}v_{Te}^2}\right) E = \frac{E}{\bar{E}}$$

We replace $E$ with $E = \bar{E}\tilde{E} = \left(\frac{m_{e}v_{Te}^2}{|q_{e}|\lambda_D}\right) \tilde{E}$ so that above, the electric field term becomes:

$$ \frac{q_i}{|q_i|}\left(\frac{|q_{i}|\lambda_D}{m_{i}v_{Te}^2}\right) E =  \frac{q_i}{|q_i|}\left(\frac{|q_{i}|\lambda_D}{m_{i}v_{Te}^2}\right) \left(\frac{m_{e}v_{Te}^2}{|q_{e}|\lambda_D}\right)\tilde{E} = \underbrace{\frac{q_i}{|q_i|}}_{=\, 1}\underbrace{\left(\frac{|q_i|}{|q_e|}\right)}_{=\, 1}\underbrace{\left(\frac{\lambda_D}{\lambda_D}\right)}_{=\, 1}\underbrace{\left(\frac{v_{Te}^2}{v_{Te}^2}\right)}_{=\, 1}\left(\frac{m_e}{m_i}\right)\tilde{E} = \frac{m_e}{m_e}\tilde{E} = \frac{1}{\mu}\tilde{E}$$

Where $\mu = m_i / m_e \simeq 1836$ has been defined. Finaly, relabelling $\tilde{E} \rightarrow E$ as before with the understanding that it is a normalized quantity, the ion species equation is found to be:

$$\boxed{\partial_t f_{i} + v\partial_x f_{i} + \frac{1}{\mu}E\partial_v f_i = 0} \qquad \underline{\text{Normalized (singly charged) ion Vlasov equation}}$$

# B.2 Incoming distribution functions at the boundary

Suppose we are interested in having Maxwellian distributions to describe our charge species entering from one side of the numerical domain. The associated normalizations from using the above need to be casted in terms of these distribution functions, i.e. suppose the normalizations have been chosen according to a species $\alpha$, then distribution functions for another species $\beta$ (or same $\beta = \alpha$), needs to be recasted consistently:

$$f_{\beta}(t,x,v) \mapsto f_{\alpha}(\omega_{p\alpha}^{-1}t, \lambda_D x, v_{T\alpha} v)$$

Take an example of an unnormalized Maxwellian centered at $v = 0$ (we save normalization until after the final form is obtained to evade leftover terms like individual masses in the prefactor whereas we want to instead use ratios)

$$f_{\beta}(t,x,v) = \exp \left(-\frac{m_{\beta}v^2}{2T_{\beta}}\right)$$

Putting this in terms of the normalized variables we have $v \mapsto v_{T\alpha} v = \sqrt{T_{\alpha} / m_{\alpha}} v$ so that the measurement is that velocity is multiples of the $\alpha$ thermal velocity.

$$f_{\beta}(t,x,v) = \exp \left(-\frac{m_{\beta}v_{T\alpha}^2 v^2}{2T_{\beta}}\right) =\exp \left(-\frac{1}{2}\frac{m_{\beta}}{m_{\alpha}}\frac{T_{\alpha}}{T_{\beta}}v^2\right) = \exp \left(-\frac{1}{2}\frac{m_{\beta}}{m_{\alpha}}\frac{T_{\alpha}}{T_{\beta}}v^2\right)$$

Define the mass ratio $\mu = m_{\beta} / m_{\alpha}$ (i.e. just as before we define the factor to be such that the species that sets the physical scales, $\alpha$ is in the denominator, and define the temperature ratio $\tau = T_{\beta} / T_{\alpha}$, the above is equivalent to:

$$f_{\beta}(t,x,v) = \exp \left(-\frac{1}{2}\frac{\mu}{\tau}v^2\right)$$

For $\beta = \alpha$, we have the unnormalized function

$$f_{\alpha}(t,x,v) = \exp \left(-\frac{1}{2}v^2\right)$$

We can normalize both functions over the infinite domain easily. For the $\alpha$ distribution, we prepend a factor of $1 / \sqrt{2\pi}$

$$\boxed{f_{\alpha}(t,x,v) = \frac{1}{\sqrt{2\pi}}\exp \left(-\frac{1}{2}v^2\right)}$$

For the $\beta$ distribution we prepend a factor of $\sqrt{\mu / \tau}$ on top of the $1 / \sqrt{2\pi}$ term:

$$\boxed{f_{\beta}(t,x,v) = \frac{1}{2\pi \tau / \mu} \exp \left(-\frac{1}{2}\frac{\mu}{\tau}v^2\right)}$$

Note, to shift the velocity distribution, we do so by the normalized coordinate $\tilde{v}$ (unlabelled in the boxed equations), where $\tilde{v} = v / v_{T\alpha}$. In the above, shifting the distribution by $n\in\mathbb{Z}$ corresponds to shifting by a multiple of the thermal velocity $v_{T\alpha}$ in both of the boxed equations.

A special case of interest because of what appears in the Bohm criterion is shifting by the sound speed:

$$c_s \simeq \sqrt{\frac{T_{\alpha}}{m_{\beta}}} = \sqrt{\frac{T_{\alpha}}{m_{\alpha}}\frac{m_{\alpha}}{m_{\beta}}} = \frac{v_{T\alpha}}{\sqrt{\mu}}$$

So the normalized quantity we add is some $\tilde{c}_s$,

$$\tilde{c}_s = \frac{c_s}{v_{T\alpha}} = \frac{1}{\sqrt{\mu}}$$

So, shifting any distribution $\beta$ by some multiple $n\in\mathbb{R}$ of the plasma sound velocity is done by:

$$f_{\beta}(t,x,v) = \frac{1}{2\pi \tau / \mu} \exp \left(-\frac{1}{2}\frac{\mu}{\tau}(v - n\tilde{c}_s)^2\right), \qquad \tilde{c}_s = \frac{1}{\sqrt{\mu}}, n\in\mathbb{R}\qquad \underline{\text{Shifted by a multiple of the sound speed in normalized variables}}$$

In general, we can of course always write

$$f_{\beta}(t,x,v) = \frac{1}{2\pi \tau / \mu} \exp \left(-\frac{1}{2}\frac{\mu}{\tau}(v - v_0)^2\right), \qquad v_0 =v / v_{T\alpha} \quad \underline{\text{in normalized coordinates for arbitrary } v_0\in\mathbb{R}}$$

# C. Setup

The simulation can occur in the presence of cold background ions. The requirement that is needed to establish a sheath is the absorbing boundary ($f_e(x = x_{w}, v_{x,outgoing}) = 0$, where $x_w < 0$ is the location of the absorbing wall. The electrons "escape" from the system on one end, leaving only positive charges. The wall should be permitted to charge up according to the electron fluxes hitting the wall. This provides a feedback into the system so that less energetic electrons will be repelled and the positive space charge can be developed. For our problem, suppose we have a left wall at $x = x_{w}$ and the bulk plasma is located at a coordinate $x = 0$. Thus we consider the evolution of electron and ion distribution functions alongside a potential $\phi : E = -\partial_x\phi$:

$$\partial_t f_e + v_x\partial_x f_e + \partial_x\phi \partial_{v_x} f_e = 0, \qquad\qquad x\in [-x_w, 0], t\in [0,T], \qquad x_w,T\in\mathbb{R}$$


$$\partial_t f_i + v_x\partial_x f_i - \frac{1}{\mu}\partial_x\phi \partial_{v_x} f_i = 0 \qquad \qquad x\in [-x_w, 0], t\in [0,T], \qquad x_w,T\in\mathbb{R}, \quad \mu = m_i / m_e$$

for

<u>Boundary conditions on density</u>

<ol>
<li>$f_{e,i}(t, -x_{w}, v_x > 0) = 0$
$$\phantom{white space for spacing}$$
<li>$f_{e,i}(t, 0, v_x) = \begin{cases} \text{chosen distribution of incoming charged species from bulk plasma} & v_x < 0 \\
&\\
0 & v_x \geq 0\end{cases}$
</ol>

<u>Boundary conditions on potential</u>

<ol>
<li> $\partial_x\phi (t, -x_{w}) = -\frac{1}{2}\sigma (t, -x_w)$
$$\phantom{white space for spacing}$$
<li> $\phi (t, 0) = 0$

Where, the surface charge density (recall our normalized units have $q_e = -1$, $q_i = +1$):

$$\sigma (t, -x_w) = \int_0^t dt' \int_{-\infty}^0 dv_x v_x\left(f_i(t', -x_w) - f_e(t', -x_{w})\right)$$

The choice of the potential in the bulk plasma to be at $\phi = 0$ implies we are interpreting the potential measurements as relative to a non-zero plasma potential. As concerns the potential at the wall, note that we do not have units such that $\epsilon_0 \rightarrow 1$ given the above development, hence we elect to choose this at this moment.

Note, since the generalized Bohm criterion cannot be checked for distributions with nonempty populations at $v_x = 0$, simulations may be warranted with truncated distributions as boundary conditions may be worth considering so that the numerical results can be directly checked with the Bohm criterion.
