# A. Introduction

Sheath problem develops from the electrostatics influence on local plasma species populations in the vicinity of a sink (boundary). The sheath itself can be collisional or collisionless, DC or RF, with mobile charged species and even are possible for coarse approximations such as cold ions provided the wall is biased significantly.

The laundry list of sheath problems we regard are as follows:

<ol>
<li> <u>matrix sheath</u> [1, p.204]: mobile electrons evolved on Vlasov-Poisson system with cold ion background, large bias voltage on boundary (wall). E.g. equivalent circumstances PIII [1, p.636], ELM bursts, any time a rush of electron density is vacated from the main plasma leaving behind regions of more populated colder ions.
$$\phantom{whitespace}$$
<li> <u>collisionless DC sheath</u>: self-consistent ion and electron Vlasov-Poisson solutions solution with boundary
$$\phantom{whitespace}$$
<li> <u>collisionless RF sheath</u> [1, p. 400]: self-consistent ion and electron Vlasov-Poisson solutions with boundary with applied field frequency satisfying
$$\omega_{pe} \gg \omega \gg \omega_{pi}$$
$${}$$
<li> <u>collisional DC sheath</u>: self-consistent ion and electron Boltzmann-Poisson solution with boundary
</ol>

We regard the formation of a positive space charge within several Debye lengths of the boundary ("edge", $n_i > n_e$). A fluid analysis produces the criterion that the ion velocity entering the "sheath" are accelerated to the sound speed ($v_{i,se} = c_i$). By considering the dynamics from the bulk plasma looking towards the edge ($\ell_{\nabla} \gg 1$), and from the edge looking outboard ($\ell_{\nabla} \ll 1$), the solutions for both scales are matched in a limit sense requiring not an inequality, but an equality. In practice, the rough agreement that ions enter the sheath supersonically is commonly observed within margin. The criterion has gained momentum over the years to be accepted as a rule-of-thumb regarding the nonlinear sheath physics. More dangerously, the Bohm criterion has become accepted to the point where shortcuts in simulations are afforded by including it as a "boundary condition" when simulating the edge. The convenience is clear, requiring supersonic velocities of ions at the sheath boundary of a simulation domain saves significant computational expense of the complicated presheath physics in order to (what is hoped) to arrive at this same result (the Bohm criterion). The sticking issue that this practice is often done yet it is acknowledged that the Bohm criterion appears to be less tenable than it was at its inception. It is necesary to create a kinetic sheath in simulation order to see how well this criterion is upheld. And, it will be seen that even this generalized Bohm criterion shows fundamental vulnerabilities in that it restricts the classes of distributions that should give rise to sheaths due to a singularity present at zero velocity should the distribution be nonempty at this value.

## A.1 Fluid derivation (classic Bohm criterion)

### Required fluid equations

The fluid equations follow as moments of the kinetic equation. Here, we take the Vlasov equation as the model.

#### Zeroeth moment: continuity equation

The classic Bohm criterion follows from the moment-based (fluid) approach, we consider the collisionless boundary problem modelling the electrons in the (adiabatic) limit. Begin with the kinetic (Vlasov) equation for a species $\alpha$:

$$\partial_t f_{\alpha} + \vec{v}\cdot \vec{\nabla}_xf_{\alpha}  + \frac{q_{\alpha}\vec{E}}{m_{\alpha}}\cdot\vec{\nabla}_vf_{\alpha} = 0$$

A mass continuity equation is obtained by integrating over $\int d^3 v$, we have

$$\int d^3 v \, \partial_t f_{\alpha} = \partial_t \int d^3 \, f_{\alpha} = \partial_t n_{\alpha}$$

$$\int d^3 \, v \vec{v}\cdot\vec{\nabla}_xf_{\alpha} = \int d^3 \, [\vec{\nabla}_x\cdot (\vec{v}f_{\alpha}) - f_{\alpha}\vec{\nabla}_x\cdot \vec{v}] = \int d^3 \, \vec{\nabla}_x\cdot (\vec{v}f_{\alpha}) =\ \vec{\nabla}_x\cdot\int d^3v \vec{v}f_{\alpha}$$

Since $\vec{\nabla}_x \cdot\vec{v} = 0$, and

$$\int d^3 v \,\frac{q_{\alpha}E}{m_{\alpha}}\vec{\nabla}_v\cdot f_{\alpha} = \frac{q_{\alpha}E}{m_{\alpha}}\cdot \int d^3 v \,\vec{\nabla}_v f_{\alpha} = 0$$

we can justify that several ways, e.g. by citing gradient theorem. This leads to a continuity equation:

$$\boxed{\partial_t n_{\alpha} + \vec{\nabla}_x\cdot (n_{\alpha}\vec{V}) = 0} \qquad \underline{\text{sourceless continuity equation}}$$

Note, if we considered a non-zero inhomogenity (source term, e.g. production terms such as ionization or other nuclear reactions). The inclusion of such a term can be blanket represented as:

$$\partial_t n_{\alpha} + \vec{\nabla}_x\cdot (n_{\alpha}\vec{V}) = S, \qquad \text{continuity equation with a source } S$$

Where $S$ is the moment of the source term in the inhomogeneous Vlasov equation (Boltzmann equation). 

#### First moment: momentum equation

Taking the first moment ($\int d^3v\,\vec{v}$) of the Vlasov equation:

$$\int d^3 v\, \vec{v} \left\{\partial_t f_{\alpha} + \vec{v}\cdot \vec{\nabla}_xf_{\alpha}  + \frac{q_{\alpha}\vec{E}}{m_{\alpha}}\cdot\vec{\nabla}_vf_{\alpha} = 0\right\}$$

we identify the required averages $n_{\alpha} \vec{V}:= \int d^3v \vec{v} f_{\alpha}$, where $\vec{V}$ is the average velocity. The third term can be written as:

$$\int d^3v \vec{v} \frac{q_{\alpha}\vec{E}}{m_{\alpha}}\cdot\vec{\nabla}_vf_{\alpha} = \frac{q_{\alpha}\vec{E}}{m_{\alpha}}\cdot\int d^3v \vec{v} \vec{\nabla}_vf_{\alpha} =  \frac{q_{\alpha}\vec{E}}{m_e}\cdot\int d^3 v \left[\vec{\nabla}_v(\vec{v} f_{\alpha}) - (\vec{\nabla}_v \vec{v})f_{\alpha}\right]$$

The first term on the right-hand side $\int_{V} d^3 v \vec{\nabla}_v\cdot (\vec{v}f_{\alpha}) = \int_{\partial S} d^2\vec{v}\cdot\vec{v}f_{\alpha} \rightarrow 0$ at $|v| \rightarrow \infty$. The second term is obvious, note that:

$$\vec{\nabla}_v \vec{v} = \partial_{v_i} v_j \hat{x}_i\hat{x}_j \delta_{ij} = \hat{x}_i\hat{x}_i =\vec{\vec{I}}$$

so the above becomes:

$$\int d^3v \vec{v} \frac{q_{\alpha}\vec{E}}{m_{\alpha}}\cdot\vec{\nabla}_vf_{\alpha} = -\frac{q_{\alpha}\vec{E}}{m_{\alpha}}\cdot\int d^3 v \vec{\vec{I}} f_{\alpha} = -\frac{q_{\alpha}\vec{E}}{m_{\alpha}}\int d^3 v f_{\alpha} = -\frac{q_{\alpha}\vec{E}}{m_{\alpha}} n_e$$

And, the middle term:

$$\int d^3 v \vec{v} \vec{v}\cdot \vec{\nabla}_xf_{\alpha} = \int d^3 v [\vec{\nabla}_x\cdot (\vec{v}\vec{v}f_{\alpha}) - f_{\alpha}\vec{\nabla}_x \cdot (\vec{v}\vec{v})] = \vec{\nabla}_x\cdot \int d^3 v \vec{v}\vec{v}f_{\alpha}$$

Since the final term $\vec{\nabla}_x \cdot (\vec{v}\vec{v}) = \partial_{x_i} \hat{x_i}\cdot v_jv_k\hat{x_j}\hat{x_k} = 0 \hat{x}_i\hat{x}_j$, since $\partial_{x_{\ell}} v_m = 0$ for all $\ell , m$.

It is convenient to rewrite the dyad in terms of the mean velocity $\vec{V} = \int d^3 v \vec{v}f_e$, so that physical quantities (pressures) can be readily identified. For example:

$$\vec{v}\vec{v} = (\vec{v} - \vec{V})(\vec{v} - \vec{V}) - \vec{V}\vec{V} + \vec{v}\vec{V} + \vec{V}\vec{v}$$

Then, the above is seen to be:

\begin{eqnarray*}
\int d^3 v \vec{v} \vec{v}\cdot \vec{\nabla}_xf_{\alpha} & =  &\vec{\nabla}_x\cdot \int d^3 v \vec{v}\vec{v}f_{\alpha} \\
&= &  \vec{\nabla}_x\cdot \int d^3 v [(\vec{v} - \vec{V})(\vec{v} - \vec{V}) - \vec{V}\vec{V} + \vec{v}\vec{V} + \vec{V}\vec{v}]f_{\alpha} \\
&= & \vec{\nabla}_x\cdot \int d^3 v (\vec{v} - \vec{V})(\vec{v} - \vec{V})f_{\alpha} - \vec{\nabla}_x\cdot \int d^3 v\vec{V}\vec{V}f_{\alpha} + \vec{\nabla}_x\cdot \int d^3 v\vec{v}\vec{V}f_{\alpha} + \vec{\nabla}_x\cdot \int d^3 v\vec{V}\vec{v}f_{\alpha} \\
&= & \vec{\nabla}_x\cdot \int d^3 v (\vec{v} - \vec{V})(\vec{v} - \vec{V})f_{\alpha} - \vec{\nabla}_x\cdot \vec{V}\vec{V}n_{\alpha} + \vec{\nabla}_x\cdot\vec{V}\vec{V}n_{\alpha} + \vec{\nabla}_x\cdot\vec{V}\vec{V}n_{\alpha} \\
&= & \vec{\nabla}_x\cdot \int d^3 v (\vec{v} - \vec{V})(\vec{v} - \vec{V})f_{\alpha} + (\vec{\nabla}_x\cdot\vec{V}\vec{V}n_{\alpha})\\
\int d^3 v \vec{v} \vec{v}\cdot \vec{\nabla}_xf_{\alpha} & = & \frac{1}{m_{\alpha}}\vec{\nabla}_x\cdot \vec{\vec{P}} + \vec{\nabla}_x\cdot(\vec{V}\vec{V}n_{\alpha})
\end{eqnarray*}

So the momentum equation is:

$$\frac{\partial(n_{\alpha}\vec{V})}{\partial t} + \vec{\nabla}_x\cdot(n_e\vec{V}\vec{V}) + \frac{1}{m_{\alpha}}\vec{\nabla}_x\cdot \vec{\vec{P}} - \frac{q_{\alpha}\vec{E}}{m_{\alpha}} n_{\alpha} = 0$$

Where we have the pressure tensor $\vec{\vec{P}}$. Often, this equation is arranged with intertial terms on the left:

$$\frac{\partial(n_{\alpha}\vec{V})}{\partial t} + \vec{\nabla}_x\cdot(n_{\alpha}\vec{V}\vec{V}) = -\frac{1}{m_{\alpha}}\vec{\nabla}_x\cdot \vec{\vec{P}} + \frac{q_{\alpha}\vec{E}}{m_{\alpha}} n_{\alpha}$$

At this point, applying the derivatives on the left-hand sides causes the continuity equation to unfold naturally, which allows the removal of some terms given continuity implies both terms sum to zero. This achieves the form most frequently quoted as the <b>momentum equation</b>:

$$\boxed{m_{\alpha}n_{\alpha}\frac{\partial\vec{V}}{\partial t} + m_{\alpha}n_{\alpha} \vec{V}\vec{\nabla}_x\cdot \vec{V} = -\vec{\nabla}_x\cdot \vec{\vec{P}} + n_{\alpha}q_{\alpha}\vec{E}} \qquad \underline{\text{momentum equation for species } \alpha}$$

Further, a partition of the pressure tensor is also convenient and clarifying:

$$\vec{\vec{P}} = p\vec{\vec{I}} + \vec{\vec{\Pi}}$$

where the scalar pressure $p = \frac{1}{3}\text{Tr}(\vec{\vec{P}})$.  Note, the pressure tensor itself is a second rank symmetric tensor, thus the trace is invariant under orthonormal coordinate transformations, hence a scalar pressure of the same value independent of coordinate system can always be defined the same way. We often consider simpler circumstances, for example if the pressure is isotropic, then the viscosity tensor $\vec{\vec{\Pi}}= \vec{\vec{0}}$. So, we reduce the pressure tensor in isotropic cases to be the following and its divergence accordingly becomes:

$$\vec{\vec{P}} = p\vec{\vec{I}} \Rightarrow \vec{\nabla}_x\cdot \vec{\vec{P}} = \vec{\nabla}_x\cdot p\vec{\vec{I}} = \partial_{x_i}\hat{x}_i\cdot p\hat{x}_j\hat{x_k} = \partial_{x_i} p \hat{x}_i = \vec{\nabla}_xp$$

Thus,

$$\boxed{\frac{\partial(m_{\alpha}n_{\alpha}\vec{V})}{\partial t} + \vec{\nabla}_x\cdot(m_{\alpha}n_{\alpha}\vec{V}\vec{V}) = -\vec{\nabla}_x p + q_{\alpha} n_{\alpha} \vec{E}} \qquad \underline{\text{isotropic pressure momentum equation for species }\alpha}$$

#### Boltzmann (adiabatic) relation for electrons 

Adiabatic in this context means slowly varying, so that the inertial terms on the left can be discarded and we consider the steady state of the electron species for the momentum equation above $q_e = -e$:

$$0 = -\vec{\nabla}_x p -  e\vec{E}n_e$$

Note, this holds true so long as there is no drift velocity. Had we included more physics, e.g. magnetic field, friction, these terms all go to zero in the special case of zero drift velocity. Hence, this result is more robust than we have let on.

With this identification, we also either have said we are neglecting magnetic fields, or that we are considering the dynamics along a field line. Both circumstances leave only electrostatic interactions to control the dynamics. Thus, the usual electric potential is afforded by virtue of the form of Faraday's law: $\vec{E} = -\vec{\nabla}_x \phi$. An equation of state can also be asserted for our sufficiently regular circumstances: $p = n_eT_e$. In general, this fluid parameter (an average) is a function of the configuration variable $x$, here we are considering a limited case where the electrons respond only to electrostatic interactions. The electrons being sufficiently mobile respond rapidly to any local variations that develop and act quickly to regularize any charge deviations as best they can (effectively accomplished in the bulk [quasineutrality], partially accomplished in the thin edge region). Notwithstanding, the temperature gradient scale length is much larger than edge characteristic scale lengths. Hence an isothermal enforcement is appropriate, and not inaccurate in the thin edge. Hence, 

\begin{eqnarray*}
\vec{\nabla}_x (n_eT_e) - en_e\vec{\nabla}_x\phi & = & 0 \\
n_e T_e\vec{\nabla}_x\ln n_e - en_e\vec{\nabla}_x\phi & = & 0 \\
n_e \vec{\nabla}_x\left(T_e\ln n_e  - e\phi\right) & = & 0 \\
\vec{\nabla}_x\left(T_e\ln n_e  - e\phi\right) & = & 0
\end{eqnarray*}

So,

$$T_e\ln n_e  - e\phi = \text{constant} \qquad \text{or} \qquad \ln n_e = \frac{\text{constant}}{T_e} + \frac{e\phi}{T_e} \Rightarrow \boxed{n_e = n_{e0}\exp\left(\frac{e\phi}{T_e}\right)} \qquad \underline{\text{Adiabatic electrons}}$$

### A.1.A Criterion from the sheath side

To find the steady state ($\partial_t\rightarrow 0$) distribution in the edge region, <b>consider a one-dimensional problem in $x$</b>. The continuity equation (no ionization should happen in the sheath) takes the form:

$$\partial_x (n_i v_i) = 0 \Rightarrow n_i = \frac{n_0 v_{i0}}{v_i}$$

Combining this with the adiabatic electron relationship into Poisson's equation (writing $n_{e0} = n_{i0} = n_0$):

$$\partial_x^2\phi = \frac{e}{\epsilon_0}(n_e - n_i) = \frac{en_0}{\epsilon_0}\left( e^{e\phi/T_e} - \frac{v_i}{v_{i0}}\right)$$

To develop an expression for $v_i / v_{i0}$, consider the isobaric momentum equation in steady state for <i>ions</i> ($q_i = +e$ for singly charged species):

$$n_iv_i\partial_x v_i = en_i E$$

which is equivalent to

$$\frac{1}{2}n_i\partial_x(v_i^2) - \frac{eE}{m_i} n_i = 0$$

or

$$n_i\frac{1}{2}\partial_x(v_i^2) = \frac{eE}{m_i} n_i = -n_i\frac{e}{m_i}\partial_x \phi$$

Which has a first integral that gives the velocity of ions:

$$v_i = \sqrt{v_{i0}^2 - \frac{2 e\phi}{m_iv_{i0}^2}} \Rightarrow \frac{v_i}{v_{i0}} = \sqrt{1 - \frac{2 e\phi}{m_iv_{i0}^2}}$$

Here, we have tacitly decided the reference potential at the sheath edge $\phi_{se} = 0$ for algebraic convenience, an inclusion of nonzero $\phi_se \neq 0$ only changes the result by replacing $\phi \rightarrow \phi - \phi_{se}$. Hence, while we lose generality in this decision, it is obvious how to incorporate it back in.

Recall that the sound speed $c_s^2 = (T_i + T_e) / m_i \simeq T_e / m_i$ for our collisionless situation since we expect nothing but $T_e \gg T_i$ (i.e. the electrons gain energy more easily than the ions from the field given their high mobility, so that $T_e \sim \frac{m_i}{m_e}T_i$ as usual, where $m_i / m_e \sim 1836$. Define the Mach number as the multiple that the peculiar ion speed  at the boundary condition (at some $x = x_{se}$, where $se$ denotes a sheath edge, i.e. the transition from the bulk plasma to the edge regions) is larger than the sound speed.

Then, we can write:

$$M_0^2 := \frac{v_{i0}}{c_s}^2 =  \frac{v_{i0}^2}{T_e / m_i} \Rightarrow v_{i0}^2 = \frac{M_0^2}{T_e / m_i}$$

Combining this with the equation for the ion velocity $v_i$, eq. (1) (Poisson's equation) becomes:

$$\partial_x^2\phi = \frac{e}{\epsilon_0}(n_e - n_i) = \frac{en_0}{\epsilon_0}\left( e^{e\phi/T_e} - \sqrt{1 - \frac{2 e\phi}{T_e M_0^2}} \right) $$

Finally, consider regimes where $e\phi / T_e \ll 1$, which permits expansion:

$$\partial_x^2\phi \simeq \frac{en_0}{\epsilon_0}\left( 1 + \frac{e\phi}{T_e} - \left( 1 + \frac{1}{2} \frac{2e\phi}{T_e}\frac{1}{M_0^2}\right) \right) = \frac{e^2n_0}{\epsilon_0T_e}\phi\left(1 - \frac{1}{M_0^2}\right) + O\left(e\phi / T_e  \right)^2 = \frac{1}{\lambda_D^2}\left(1 - \frac{1}{M_0^2}\right)\phi$$

Then, we have

$$\partial_x^2\phi = \frac{\phi}{\lambda_D^2}\left(1 - \frac{1}{M_0^2}\right)$$

One of the most controversial parts of this derivation is that Bohm falls back on purely mathematical concerns to finalize the result. For non-oscillatory solutions, we require

$$1 - \frac{1}{M_0^2} \geq 0 \Rightarrow 1 \geq \frac{1}{M_0^2} \Rightarrow M_0 \geq 1$$

Recalling $M_0 := v_{i0} / c_s$, this demands

$$v_{i0} \geq c_s, \qquad \text{Bohm criterion from the sheath side in the fluid model}$$

### A.1.B From the plasma bulk side

Analyzing the situation from the bulk looking towards the edge, we arrive at almost completely contradictory criteria, yet there is exactly one solution that is compatible for both, but only in a limit sense.

Note, that for steady state analysis, we must include a source or else the plasma would be depleted as the species populations would are absorbed by the wall sink. A source is needed to replenish the plasma and produce a stable population. Hence, we use the latter form of the continuity equation above, and consider 1D steady state ($\partial_t \rightarrow 0$)

$$n\partial_x v_i + v_i \partial_x n = S$$

and consider the ion momentum equation (see boxed equation above for electrons, we take $q_i = e$ to recover the ion equation):

$$m_i nv_i \partial_x v_i = enE$$

In steady state, the electrons are to a good approximation adiabtic:

$$enE + T_e\partial_x n = 0 \Rightarrow enE = - T_e\partial_x n$$

Combining this with the ion momentum equation:

$$m_i nv_i \partial_x v_i = - T_e\partial_x n$$

Solving for the derivative of the density from the continuity equation: 

$$\partial_x n = \frac{1}{v_i}S - \frac{n}{v_i}\partial_x v_i$$

Inserting this derivative into the equation just prior:


\begin{eqnarray*}
m_i nv_i \partial_x v_i & = & - T_e\left(\frac{1}{v_i}S - \frac{n}{v_i}\partial_x v_i\right) \\
\partial_x v_i & = & -\left(\frac{T_e}{m_i}\frac{1}{nv_i^2} S - \frac{1}{m_iv_i^2}T_e \partial_x v_i\right) \\
\partial_x v_i & = & -\left(\frac{c_s^2}{nv_i^2} S - \frac{1}{m_iv_i^2}T_e \partial_x v_i\right) \\
\partial_x v_i \left(1 -  \frac{1}{m_iv_i^2}T_e\right) & = & -\frac{c_s^2}{nv_i^2}S \\
\partial_x v_i & = & -\frac{\frac{c_s^2}{nv_i^2}S}{1 -  T_e / (m_iv_i^2)} 
\end{eqnarray*}

or

$$\partial_x v_i  =  -\frac{c_s^2 S / n}{v_i^2 -  c_s^2} \qquad c_s^2 = T_e / m_i $$

The ion velocity $v_i$ increases from the bulk (small $x$ in our implicit coordinate system) from smaller values. Yet, a singularity develops exactly as it reaches the sound speed $c_s$, hence we require 

$$v_i < c_s, \qquad \text{Sheath (fluid) criterion from the bulk side looking towards the sheath}$$

Recalling the Bohm criterion from the sheath side looking out towards the bulk insisted $v_i \geq c_s$, taking the threshold limit we recover the fluid result for a collisionless plasma must have ions entering with a fluid velocity of 

$$\boxed{v_i = c_s} \qquad \text{ion fluid velocity at the sheath entrance of a collisionless plasma}$$




Many times, the condition $v_{i}\geq c_s$ is quoted as the Bohm criterion, especially since in actual experients there is a spreading of velocities entering the sheath given distributions of velocities exist rather than a single fluid velocity. This also broadens the transition region and obscures how to define a clean "sheath edge" for the plasma. It is clear that in real plasmas, if we track the point of immediate departure from quasineutrality $n_i \simeq n_e$, this point tends to always correspond significantly inboard from the wall, much further than one would normally interpret in a $n$ vs. $x$ plot as what appears to be the sheath edge. Different conventions can be adopted that are more observationally consistent, e.g. the $x$ coordinate where we have a deviation by 1%, or perhaps 10% can be considered. This is done often in literature.

#### after note:

When is $e\phi / T_e \ll 1$ representative of the plasma? Note that we can write this in terms of the Debye length $\lambda_D^2 = \epsilon_0 T_e / (n_0 e^2)$ 

$$\frac{e\phi}{T_e} = \frac{e\epsilon_0}{\lambda_D^2 n_0} \phi$$

In typical plasmas we have $\lambda_D \sim 10^{-5} - 10^{-4}$ m, the constant $\epsilon_0 \sim 10^{-12} \mathrm{m^{-3} kg^{-1} s^4 A^2}$, so it is not hard to have this condition be met. Only in extremely rareified plasmas (e.g. astronomical plasmas with length scales on correspondingly extreme lengths by everyday experiment standards) might we find this not met in anything considered a plasma. To be even more blatant about it, note that we can write this as:

$$\frac{e\phi}{T_e} = \frac{e\epsilon_0}{\lambda_D^2 n_0^{2/3}n_0^{1/3}} \phi =\frac{e\epsilon_0}{(\lambda_D^3 n_0)^{2/3}n_0^{1/3}} \phi$$

Which states this quantity is small ($\frac{e\phi}{T_e} \ll 1$) so long as $n_0\lambda_D^3 \gg 1$. That is, that there exists many particles in a Debye cube. This is part of the definition of a plasma.



# A.2 Kinetic derivation of generalized Bohm criterion

The kinetic criterion is obtained even more quickly than the fluid stepthrough. We aim to find a criterion that arises in the steady state system, just as before.

Begin with the Vlasov equation for a species $\alpha$ at steady state ($\partial_t \rightarrow 0$):

$$\vec{v}\cdot \vec{\nabla}_x f_{\alpha} + \frac{q_{\alpha}}{m_{\alpha}} \vec{E}\cdot \vec{\nabla}_v f_{\alpha} = 0$$ 

Considering any component $x_i = \{x_1, x_2, x_3\}$:

$$v_{x_i}\partial_{x_i} f_{\alpha} + \frac{q_{\alpha}}{m_{\alpha}} E_{x_i}\partial_{v_{x_i}} f_{\alpha} = 0$$ 

or 

$$\partial_{x_i} f_{\alpha} + \frac{q_{\alpha}}{m_{\alpha}} E_{x_i}\frac{1}{v_{x_i}}\partial_{v_{x_i}} f_{\alpha} = 0$$ 

integrating ($\int d^3 v$) and identifying $n_{\alpha} := \int d^3 v f_{\alpha}$

$$\partial_{x_i} n_{\alpha} + \frac{q_{\alpha}}{m_{\alpha}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}}\partial_{v_{x_i}} f_{\alpha} = 0 \qquad \qquad (1)$$ 

then,

$$\partial_{x_i} n_{\alpha} = -\frac{q_{\alpha}}{m_{\alpha}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}}\partial_{v_{x_i}} f_{\alpha} = 0$$ 


In terms of species $\alpha = e,i$:

$$\partial_{x_i} n_{i} = -\frac{q_{i}}{m_{i}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}}\partial_{v_{x_i}} f_{i}$$ 
$$\partial_{x_i} n_{e} = -\frac{q_{e}}{m_{e}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}}\partial_{v_{x_i}} f_{e}$$ 

Often, the ion equation has the last integral rewritten according to the following. Note that:

$$\frac{1}{v_{x_i}} \partial_{v_{x_i}} f_i = \partial_{v_{x_i}} \left(\frac{1}{v_{x_i}}f_i\right) + \frac{1}{v_{x_i}^2}f_i$$

so that the first integral on the right-hand side becomes

$$\int d^3v\, \partial_{v_{x_i}} \left(\frac{1}{v_{x_i}}f_i\right) = \int d^2v\left(\frac{1}{v_{x_i}}f_i\right) = 0 \qquad \text{at the
surface } |v|\rightarrow 0$$

leaving:

$$-\frac{q_{i}}{m_{i}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}}\partial_{v_{x_i}} f_{i} = -\frac{q_{i}}{m_{i}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}^2} f_{i}$$

and the first moment of the steady state ion Vlasov equation is:

$$\partial_{x_i} n_{i} = -\frac{q_{i}}{m_{i}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}^2} f_{i}$$ 

Inside the sheath we have $n_i > n_e$. In the bulk plasma, we have $n_i \simeq n_e$. The sheath edge marks the start of this deviation, i.e. at the sheath edge and inside we must have

$$n_i \geq n_e \quad \Rightarrow \quad \partial_{x_i} n_i \geq \partial_{x_i} n_e$$

Inserting the above two equations, and considering the case of singly charged ions $(q_i = e, q_e = -e$) we have (sign error has occured, the correct result is shown below):

$$\frac{e}{m_{i}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}^2} f_{i} \leq -\frac{e}{m_{e}} E_{x_i}\int d^3 v\, \frac{1}{v_{x_i}}\partial_{v_{x_i}} f_{e}$$ 

or,

$$\boxed{-\frac{1}{m_{e}}\int d^3 v\, \frac{1}{v_{x_i}}\partial_{v_{x_i}} f_{e} \geq \frac{1}{m_{i}}\int d^3 v\, \frac{1}{v_{x_i}^2} f_{i}} \qquad \text{Kinetic Bohm criterion}$$ 

### Recovering the fluid Bohm criterion

To recover the fluid approximation, we consider the "kinetic" fluid case, i.e. all ions move at the fluid velocity $V_i$, so that $f_i(v) = n_0\delta (\vec{v} - \vec{V}_i)$, also in a fluid description we are at local thermodynamic equilbrium, so that $f_e(v) = n_0 (m / 2\pi T_e)\exp (-m_ev^2 / (2 T_e))$ and $\int d^3 v f_e(v) = n_0$.

The criterion then takes the form (where we take $v_{x_i} \equiv v)$

\begin{eqnarray*}
-\frac{1}{m_{e}}\int d^3 v\, \frac{1}{v}\partial_{v} f_{e} & \geq & \frac{1}{m_{i}}\int d^3 v\, \frac{1}{v^2} f_{i} \\
-\frac{1}{m_{e}}\int d^3 v\, \frac{1}{v}\partial_{v} \left[ n_0\left(\frac{m_e}{2\pi T_e}\right)^{3/2}\exp \left(-\frac{m_ev^2}{2 T_e}\right)\right] & \geq &\frac{1}{m_{i}}\int d^3 v\, \frac{n_0}{v^2} \delta (\vec{v} - \vec{V}_i)  \\
-\frac{1}{m_{e}}\int d^3 v\, \frac{1}{v}\left(-\frac{m_ev}{T_e}\right) n_0\left(\frac{m_e}{2\pi T_e}\right)^{3/2}\exp \left(-\frac{m_ev^2}{2 T_e}\right) & \geq & \frac{n_0}{m_{i}}\frac{1}{V_i^2} \\
+\left(\frac{m_e}{T_e}\right)\frac{1}{m_{e}}\int d^3 v\, f_e(v) & \geq &\frac{n_0}{m_{i}}\frac{1}{V_i^2} \\
\left(\frac{m_e}{T_e}\right)\frac{1}{m_{e}}n_0 & \geq &\frac{n_0}{m_{i}}\frac{1}{V_i^2}\\
\frac{1}{T_e} & \geq &\frac{1}{m_{i}}\frac{1}{V_i^2} \\
\frac{1}{T_e} & \geq &\frac{1}{m_{i}}\frac{1}{V_i^2} \\
\frac{m_i}{T_e} & \geq &\frac{1}{V_i^2} \\
\frac{1}{c_s^2} & \geq &\frac{1}{V_i^2}
\end{eqnarray*}

Or,

$$\boxed{V_i \geq c_s}$$

So, kinetic criterion checks out in this way and seems to provide appropriate generalization.

## Existence of presheath

For subsonic ion velocities that exist before the sheath entrance, there must exist an electric field accelerating them to sonic speeds. This region needs to be much wider than the sheath region so that it can accelerate the majority of ions at very different initial speeds to the the sound speed. This region is called the <i>presheath</i>. It is a region over which the ions transition to supersonic speeds largely unimpeded so they have enough distance to ramp up their velocity (collisionless), and at which point (sheath entrance) the condition of charge neutrality no longer holds since the electrons are quickly lost to the sink leaving a surplus of ions.

## A.3 Note about controversy


#### on the original Bohm criterion

There is significant discussion about the claimed interpretation of the Bohm criterion as well as the tacit conditions that must be made in order to obtain his result. Bohm acknowleges this in his principal paper, though does not put the words much specificity about what is off about both the stepthrough and the resulting interpretation. Actually, it is not until the final sentence that he puts forth the interpretation widely used (read: abused) in plasma physics explanations. Bohm claims the criterion, though obtained through uncareful means seems to hold in most experimental situations within "rough[ly] 20%", i.e. that the ions do enter the sheath within the sound velocity. Thus, it is expected that a guideline like this criterion not be considered too critically, especially given it is difficult to determine where exactly in the plasma bulk-edge transition region the criterion should be met.

His final sentence reads: "Thus, it may be said roughly that a sheath begins when $V_0 = kT_e /2e$ [$V$ is the potential], remembering that this means that beyond this point the potential begins to climb rapidly and the electron density to drop very rapidly.” There are no observations that deny that this does happen with the species populations, it is the definition of a sheath edge and the climbing of the potential that does not actually follow from what Bohm obtained. To that end, this author agrees with several others in that Bohm misinterpreted his result. His solution corresponds to an asymptotic solution, whose level curve at $v_i = c_s$ (sound speed) in the natural Hamiltonian phase space ($E$, $\phi$) corresponds to an unstable manifold. Indeed, Bohm obtains a solution corresponding to the sheat being located in the limit $x\rightarrow -\infty$. He misinterprets the matching point of the two-scale analysis (bulk plasma vs. edge plasma) as being identical to the point where $V = V_0$, a reference point where $E = 0$. This isn't technically true, and can be shown straightforwardly that it deviates from this point in phase space ($E$, $\phi$) in general [2]

#### on the kinetic Bohm criterion [cf. Hegna, Callen back-and-forth in IEEE]

As shown above in section A.2, a necessary step for the kinetic criterion to be obtained is the division by a velocity parallel to the electric field (above, $v_{x_i}$). The singularities that persist have no means to be removed, so this requires that the distribution function $f_i(v_{x_i} = 0) = 0$, otherwise the right-hand side diverges at the origin in phase space, and the left-hand side integrates to a quantity proportional to the temperature normalized by a density, which clearly cannot be zero. A second thought rationalization is thet this then requires in the sheath region that there exist only one-sided velocities on the ion distribution. This is reassuring given it is the expectation in this region. If any ion escapes (threshold at $f_i(v_{x_i} \leq 0$), then this is not the sheath region. However, it is not satisfying that even basic distributions such as Maxwellian ions cannot be captured by this criterion. There is other issue that some authors contest with this rationalization as well (e.g. the kinetic description takes a continuum limit so that $\lambda_D / L \rightarrow 0$ so the sheath is infinitely thin), but as concerns the present purposes, it is sufficient to close the controversy survey here given all the main points have been outlined. One bottom line is that the kinetic criterion because of its divergenceless requirement makes the criterion experimentally untestable. The nonlinear Vlasov solver DECSKS-2.0 will provide a check on this criterion, we will see where this condition is of help, thus locating the "sheath edge." 

# B. Setup

The simulation can occur in the presence of cold background ions. The requirement that is needed to establish a sheath is the absorbing boundary ($f_e(x = x_{w}, v_{x,outgoing}) = 0$, where $x_w < 0$ is the location of the absorbing wall. The electrons "escape" from the system on one end, leaving only positive charges. The wall should be permitted to charge up according to the electron fluxes hitting the wall. This provides a feedback into the system so that less energetic electrons will be repelled and the positive space charge can be developed. For our problem, suppose we have a left wall at $x = x_{w}$ and the bulk plasma is located at a coordinate $x = 0$. Thus we consider the evolution of electron and ion distribution functions alongside a potential $\phi : E = -\partial_x\phi$ (see <a href = "https://github.com/dsirajud/IPython-notebooks/blob/master/DECSKS-14%20--%20Two_species_Vlasov_simulations/DECSKS-14%20--%20Two_species_Vlasov_simulations.ipynb">DECSKS-14, sec. B</a> for details on normalizations):

$$\partial_t f_e + v_x\partial_x f_e + \partial_x\phi \partial_{v_x} f_e = 0, \qquad\qquad x\in [-x_w, 0], t\in [0,T], \qquad x_w,T\in\mathbb{R}$$


$$\partial_t f_i + v_x\partial_x f_i - \frac{1}{\mu}\partial_x\phi \partial_{v_x} f_i = 0 \qquad \qquad x\in [-x_w, 0], t\in [0,T], \qquad x_w,T\in\mathbb{R}, \quad \mu = m_i / m_e$$

for

<u>Boundary conditions on density</u>

<ol>
<li>$f_{e,i}(t, -x_{w}, v_x > 0) = 0$
$$\phantom{white space for spacing}$$
<li>$f_{e,i}(t, 0, v_x) = \begin{cases} \text{chosen distribution of incoming charged species from bulk plasma} & v_x < 0 \\
&\\
0 & v_x \geq 0\end{cases}$
</ol>

<u>Boundary conditions on potential</u>

<ol>
<li> $\partial_x\phi (t, -x_{w}) = -\frac{1}{2}\sigma (t, -x_w)$
$$\phantom{white space for spacing}$$
<li> $\phi (t, 0) = 0$

Where, the surface charge density (recall our normalized units have $q_e = -1$, $q_i = +1$):

$$\sigma (t, -x_w) = \int_0^t dt' \int_{-\infty}^0 dv_x v_x\left(f_i(t', -x_w) - f_e(t', -x_{w})\right)$$

The choice of the potential in the bulk plasma to be at $\phi = 0$ implies we are interpreting the potential measurements as relative to a non-zero plasma potential. As concerns the potential at the wall, note that we do not have units such that $\epsilon_0 \rightarrow 1$ given the above development, hence we elect to choose this at this moment.

Note, since the generalized Bohm criterion cannot be checked for distributions with nonempty populations at $v_x = 0$, simulations may be warranted with truncated distributions as boundary conditions may be worth considering so that the numerical results can be directly checked with the Bohm criterion.
